```python
def countBinarySubstrings(self, s: str) -> int:
    res = 0
    i = 0
    pre,curr = 0,0
    while i < len(s):
        if s[i]==s[i+1]:
            curr += 1
            i+=1
            continue
        else:
            res += min(pre,curr)
            pre = curr
            curr = 0
            i+=1
    return res
        
```
