# 题目**1022**✨

```python
from typing import Optional
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def sumRootToLeaf(self, root: Optional[TreeNode]) -> int:
    res = 0
    def dfs(root:Optional[TreeNode],curr:str)->None:
        nonlocal res
        if not root.left and not root.right:
            curr+=str(root.val)
            res+=int(curr,2)
            return
        temp = curr.add(str(root.val))
        if root.left:
            dfs(root.left,curr)
        if root.right:
            dfs(root.right,curr)
    dfs(root,"")
    return  res
```

# 题目**22**✨✨

```python
from typing import List
```

```python
def generateParenthesis(self, n: int) -> List[str]:
    res = []
    def backtracking(curr:str,left:int,right:int):
        if left ==0 and right ==0:
            res.append(curr)
            return
        if right>left:
            return
        if left>0:
            backtracking(curr+"(",left-1,right)
        if right>0:
            backtracking(curr+")",left,right-1)
    backtracking("",n,n)
    return res


```

# 题目**79**✨✨

```python
def exist(self, board: List[List[str]], word: str) -> bool:
    m = len(board)
    n = len(board[0])
    visited = [[False]*n for _ in range(m)]
    def dfs(i: int,j:int,nums:int)->bool:
        nonlocal visited
        if board[i][j] != word[nums] or visited[i][j]:
            return  False
        if nums == len(word)-1:
            return  True
        if board[i][j] == word[nums] and not visited[i][j]:
            visited[i][j] = True
            for x,y in {(1,0),(-1,0),(0,-1),(0,1)}:
                xi,yj = i+x,j+y
                if 0<=xi<m and 0<=yj<n and dfs(xi,yj,nums+1):
                    return True
            visited[i][j] = False
        return False

    for i in range(m):
        for j in range(n):
            if dfs(i,j,0):
                return True
    return  False

```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    if head is None or head.next is None:
            return head
    sec = self.get_next_mid(head)
    one = self.sortList(head)
    two = self.sortList(sec)
    return self.mergeTwoList(one,two)

    return None
def mergeTwoList(self,one:Optional[ListNode],two:Optional[ListNode])->None:
    dummy = ListNode(-1)
    start = dummy
    while one and two:
        if one.val<two.val:
            start.next = one
            one = one.next
        else:
            start.next = two
            two = two.next
        start = start.next
    start.next = two if two else one
    return dummy.next
def get_next_mid(self,head:Optional[ListNode])->Optional[ListNode]:
    slow,fast,pre = head
    while fast and fast.next:
        pre = slow
        slow = slow.next
        fast = fast.next.next
    pre.next = None
    return slow
```
