# 题目**3296**✨✨✍️

```python
from typing import List
from math import isqrt
```

```python
def minNumberOfSeconds(self, mountainHeight: int, workerTimes: List[int]) -> int:
    left = 0,right = min(workerTimes)*mountainHeight*(mountainHeight+1)//2+1
    canHeight = 0
    while left<right:
        mid = left+(right-left)//2
        canHeight = 0
        for t in workerTimes:
            k = (-t+isqrt(t**2+8*mid))//(2*t)
            canHeight +=k
        if canHeight>=mountainHeight:
            right = mid
        else:
            left  = mid+1
    return left
```

# 题目**12**✨✨

```python
def intToRoman(self, num: int) -> str:
    res = ""
    temp = num
    count = -1
    hashtable = {
        1:"I",
        5:"V",
        10:"X",
        50:"L",
        100:"C",
        500:"D",
        1000:"M",
        4:"IV",
        9:"IX",
        40:"XL",
        90:"XC",
        400:"CD",
        900:"CM"
    }
    while temp != 0:
        a = temp % 10
        temp = temp//10
        count+=1
        if a<4:
            while a !=0:
                a-=1
                res = hashtable[1*(10**count)]+res
        elif a ==4:
            res = hashtable[4*(10**count)]+res
        elif a ==5:
            res = hashtable[5*(10**count)]+res
        elif a== 9:
            res = hashtable[9*(10**count)]+res
        else:
            a-=5
            while a !=0:
                a-=1
                res = hashtable[1*(10**count)]+res
            res = hashtable[5*(10**count)]+res
    return res 
```

## 方法二，更快速的方式，时间复杂度低的

```python
def intToRoman(self, num: int) -> str:
    VALUE_SYMBOLS = [
        (1000, "M"),
        (900, "CM"),
        (500, "D"),
        (400, "CD"),
        (100, "C"),
        (90, "XC"),
        (50, "L"),
        (40, "XL"),
        (10, "X"),
        (9, "IX"),
        (5, "V"),
        (4, "IV"),
        (1, "I"),
    ]
    roman = list()
    for value,symbol in VALUE_SYMBOLS:
        while num>=value:
            num-=value
            roman.append(symbol)
        if num == 0:
            break
    return "".join(roman)
```

# 题目**76**✨✨✨

```python
from collections import defaultdict
```

```python
def minWindow(self, s: str, t: str) -> str:
    count = defaultdict(int)
    for i in t:
        count[i]+=1
    total = len(count)

    left = 0
    ansleft = -1
    ansright = len(s)

    for right,ss in enumerate(s):
        if ss in count:
            count[ss]-=1
            # 这个得是在ss在目标范围内的情况下才可以，否则不能这么操作
            if count[ss]==0:
                total-=1
        while total ==0:
            if ansright-ansleft>right-left:
                ansleft,ansright = left,right
            if s[left] in count:    
                if count[s[left]]==0:
                    total+=1
                count[s[left]]+=1
            left+=1


    return "" if ansleft == -1 else s[ansleft:ansright+1]

```

# 题目**36**✨✨

```python
def isValidSudoku(self, board: List[List[str]]) -> bool:
    row =set()
    col = set()
    subbox = set()
    for i in range(9):
        for j in range(9):
            x = board[i][j]
            if x == ".":
                continue
            if (i,x) in row or (j,x) in col or (i//3,j//3,x) in subbox:
                return False
            row.add((i,x))
            col.add((j,x))
            subbox.add((i,j,x))
    return True
```

# 题目**23**✨✨✨

```python

def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
    def mergeTwoLists(list1:Optional[ListNode],list2:Optional[ListNode])->Optional[ListNode]:
        dummy = ListNode(-1)
        temp = dummy
        while list1 and list2:
            if list1.val<list2.val:
                temp.next = list1
                list1 = list1.next
            else:
                temp.next = list2
                list2 = list2.next
            temp = temp.next
        temp.next = list1 if list1 else list2
        return dummy
            
    if not lists:
        return None
    while len(lists)>1:
        merged = []
        for i in range(len(lists),2):
            if i+1<len(lists):
                merged.append(mergeTwoLists(lists[i],lists[i+1]))
            else:
                merged.append(lists[i])
        lists = merged
    return lists
        
```

# 题目**912**✨✨

```python
def sortArray(self, nums: List[int]) -> List[int]:
    def merge(left:int,right:int):
        if left>=right:
            return 
        mid = left+(right-left)//2
        merge(left,mid)
        merge(mid+1,right)
        i,j = left,mid+1
        temp = []
        while i<=mid and j<=right:
            if nums[i]<nums[j]:
                temp.append(nums[i])
                i+=1
            else:
                temp.append(nums[j])
                j+=1
        if i<=mid:
            temp+=nums[i:mid+1]
        else:
            temp+=nums[j:right+1]

        nums[left:right+1] = temp
    merge(0,len(nums)-1)
    return nums

```
