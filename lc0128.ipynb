>整体是按照动态规划的方式去进行处理，然后每次在遍历的时候，分为两个部分：  
>1. 按照没有传送的方式去进行更新，就是使用传统的动态规划
>cost[i][j] = min(cost[i+1][j]+grid[i+1][j],cost[i][j+1]+grid[i][j+1])
>2. 按照传送的更新方式，这种方式要求对于grid[x][y]<=grid[i][j]
>  
>这两种方式是同时发挥作用的，先进行2方式，然后再进行1方式
>

```python
from typing import List
```

```python
def minCost(self, grid: List[List[int]], k: int) -> int:
    # 构建points数组，存储(i,j)，然后按照grid的值进行升序排序
    m,n = len(grid),len(grid[0])
    points = [(i,j) for i in range(m) for j in range(n)]
    points.sort(key= lambda p :grid[p[0]][p[1]])
    
    # 构建cost数组，用来存储当前到达终点所需要的cost
    costs = [[float('inf')]*n for _ in range(m)]

    for t in range(k+1):
        # 先进行方式2的更新
        minCost = float('inf')
        j = 0
        for i in range(len(points)):
            minCost = min(minCost,costs[points[i][0]][points[i][1]])
            if i+1<len(points) and grid[points[i][0]][points[i][1]] == grid[points[i+1][0]][points[i+1][1]]:
                # i+=1
                # 在这个地方中，修改i是没有用的。
                # 后面再迭代的时候，i只会通过全局变量将这个部分再调整一次
                continue
            for r in range(j,i+1):
                costs[points[r][0]][points[r][1]] = minCost
            j = i+1
        
        # 进行方式1的更新
        for i in range(m-1,-1,-1):
            for j in range(n-1,-1,-1):
                if i ==m-1 and j ==n-1:
                    costs[i][j] =0
                    continue
                if i != m-1:
                    costs[i][j] = min(costs[i+1][j]+grid[i+1][j],costs[i][j])
                if j != n-1:
                    costs[i][j] = min(costs[i][j+1]+grid[i][j+1],costs[i][j])
    
    return costs[0][0]

```

```python
def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
        # 对于本题而言，使用类似插入排序的方式
        for i in range(n):
                nums1[m+i] = nums2[i]
        
        for i in range(m,m+n-1):
                temp = nums1[i]
                loc = i
                for j in range(i,0,-1):
                        if nums1[j-1]>nums1[j]:
                                nums1[j]= nums1[j-1]
                        else:
                                loc = j
                                break
                nums1[loc]= temp
```

>用类似于归并排序的最后一步的操作去处理  
>这两个有序的数组

```python
def merge(self, nums1: List[int], m: int, nums2: List[int], n: int) -> None:
    p1,p2 =0,0
    sorted= []
    while p1<m and p2<n:
        if nums1[p1]<nums2[p2]:
            sorted.append(nums1[p1])
            p1+=1
        else:
            sorted.append(nums2[p2])
            p2+=1
    if p2 ==n:
        sorted+=nums1[p1:m] 
    else:
        sorted += nums2[p2:n]
    nums1[:] = sorted

```

```python
def removeElement(self, nums: List[int], val: int) -> int:
    # 采用双指针，一个用来遍历当前所处的需要被替换掉的数据
    # 一个指向正常的数值，用来和这个要替换的值进行对调
    left = 0
    right = 0
    while right<len(nums):
        if nums[right]!=val:
            nums[left]=nums[right]
            left +=1
        
        right+=1

    return left
```

```python
def removeDuplicates(self, nums: List[int]) -> int:
    left,right =0,1
    while right <len(nums):
        if nums[right]==nums[left]:
            right+=1
        else:
            left+=1
            nums[left]=nums[right]
    return left+1

```
