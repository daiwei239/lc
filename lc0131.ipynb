# 题目**744**✨

>本题直接做，或者采用二分的方式

```python
def nextGreatestLetter(self, letters: List[str], target: str) -> str:
    if target>=letters[-1]:
        return letters[0]
    n = len(letters)
    left,right = 0,n-1
    while left<right:
        mid = (left+right)//2
        if letters[mid]<=target:
            left = mid+1
        else:
            right = mid        
    return letters[left] 
        
```

# 题目**55**✨✨

>使用贪心的方式求解

```python
from typing import List
```

```python
def canJump(self, nums: List[int]) -> bool:
    n=len(nums)
    max_len = nums[0]
    for i in range(0,n):
        if i<=max_len:
            max_len=max(max_len,i+nums[i])
    return max_len>=(n-1)
```

# 题目**45**✨✨

>采用贪心算法直接去做就可以了

```python
def jump(self, nums: List[int]) -> int:
    n = len(nums)
    step = 0
    maxpos = 0
    end =0

    for i in range(n):
        if i<=end:
            maxpos = max(maxpos,i+nums[i])
        if i==end:
            end = maxpos
            step+=1
    return step
```

# 题目**274**✨✨

>直接用遍历法做，时间复杂度比较高

```python
def hIndex(self, citations: List[int]) -> int:
    # 先对引用进行排序，然后遍历整个数组，
    # 比较n-i和citation[i]的值的大小，
    # 如n-i>citation[i]
    # 即有n-i篇论文的引用量大于citation[i]
    # 要求到的目标是max{citation[i]}
    n = len(citations)
    h = 0
    citations.sort()
    for i in range(n):
        if n-i>citations[i]:
            h = citations[i]
    return h
```
