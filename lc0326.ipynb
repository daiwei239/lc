# 题目**3548**✨✨

```python
from typing import List
```

```python
def canPartitionGrid(self, grid: List[List[int]]) -> bool:
    total = sum(sum(row) for row in grid)
    def check(a:List[List[int]])->bool:
        m = len(a)
        n = len(a[0])
        # 删除上半段的数，是否可以满足对应要求
        def f(a:List[List[int]])->bool:
            stack = {0}
            currSum = 0
            for i,row in enumerate(a[:-1]):
                for j,x in enumerate(row):
                    currSum+=x
                    if i>0 or j==0 or j == n-1:
                        stack.add(x)
                if n ==1:
                    if currSum * 2 == total or currSum *2 -total == a[0][0] or currSum*2-total == row[0]:
                        return True
                    continue
                if currSum *2 -total in stack:
                    return True
                if i == 0:
                    stack.update(row)
            return False
        return f(a) or f(a[::-1])
    return check(grid) or check(list(zip(*grid)))            
```

# 题目**560**✨✨

```python
from typing import collections,defaultdict
```

```python
def subarraySum(self, nums: List[int], k: int) -> int:
    
    if not nums:
        return 0
    cnt = defaultdict(int)
    cnt[0] = 1
    res = 0
    prefix = 0
    for x in nums:
        prefix+=x
        if prefix-k in cnt:
            res+=cnt[prefix-k]
        cnt[prefix]+=1
    return res

```

## 方法二，自己原本写的代码       

```python
def subarraySum(self, nums: List[int], k: int) -> int:
        hash =  defaultdict(set)
        hash[nums[0]].add(0)
        hash[0].add(-1)
        res = 0
        if nums[0]==k:
            res =1
        for i in range(1,len(nums)):
            nums[i] +=nums[i-1]
            if nums[i]-k in hash:
                res+=len(hash[nums[i]-k])
            hash[nums[i]].add(i)

        return res
            
```

# 题目**34**✨✨

```python
def searchRange(self, nums: List[int], target: int) -> List[int]:
    def find_position(left:int,right:int,k:int)->int:
        while left<=right:
            mid = left+(right-left)//2
            if nums[mid]>k:
                right = mid-1
            else:
                left = mid+1
        return left-1





    if not nums:
        return [-1,-1]
    one = find_position(0,len(nums)-1,target-1)
    two = find_position(0,len(nums)-1,target)
    if nums[two]!= target:
        return [-1,-1]
    if nums[one] == target-1:
        return [one+1,two]
    return[one,two]
    
```

# 题目**20**✨

```python
from collections import deque
```

```python
def isValid(self, s: str) -> bool:
    stack= deque()
    for ch in s:
        if ch in ['(',"{","["]:
            stack.appendleft(ch)
            continue
        if ch ==')':
            if stack[0] == '(':
                stack.popleft()
            else:
                return False
        elif ch ==']':
            if stack[0] == '[':
                stack.popleft()
            else:
                return False
        elif ch =='}':
            if stack[0] =='{':
                stack.popleft()
            else:
                return False
    return len(stack)==0
```
