```python
from typing import List
```

```python
class UnionFind:
    def __init__(self,n:int):
        self._fa = list(range(n))
        self.cc = n

    def find(self,x:int)->int:
        if self._fa[x]!=x:
            self._fa[x] = self.find(self._fa[x])
        return self._fa[x]
    
    def merge(self,from_:int,to_:int)->bool:
        x,y = self.find(from_),self.find(to_)
        if x == y:
            return False
        self._fa[x] = y
        self.cc -=1
        return True
    
class Solution:
    def maxStability(self,n:int,edges:List[List[int]],k:int)->int:
        must_uf = UnionFind(n)
        all_uf = UnionFind(n)
        min_s,max_s = inf,0
        for x,y,s,must in edges:
            if must and not must_uf.merge(x,y):
                return -1
            
            all_uf.merge(x,y)
            min_s = min(min_s,s)
            max_s = max(max_s,s)
        if all_uf.cc>1:
            return -1
    
        def check(low:int)->bool:
            u = UnionFind(n)

            for x,y,s,must in edges:
                if must and s<low:
                    return False
                if must or s>=low:
                    u.merge(x,y)
            left_k = k
            for x,y,s,must in edges:
                if left_k == 0 or u.cc ==1:
                    break
                if not must and s<low<=s*2 and u.merge(x,y):
                    left_k -=1
            return u.cc ==1

        left,right = min_s,max_s*2+1
        while left+1 < right:
            mid = (left+right)//2
            if check(mid):
                left = mid
            else:
                right = mid
        return left
    
```

# 题目**30**✨✨✨

```python
from typing import Counter
from collections import defaultdict
```

```python
# 采用滑动窗口的线性解法
def findSubstring(self, s: str, words: List[str]) -> List[int]:
    slen = len(s)
    llen = len(words[0])
    wlen = len(words)*llen
    if slen<wlen:
        return []
    target = Counter(words)   # 修改：记录 words 中每个单词需要出现的次数
    res = []
    # 将单词进行分组进行滑动窗口
    for i in range(llen):
        left = i
        count = 0

        seen = defaultdict(int)

        for right in range(i,slen-llen+1,llen):
            word = s[right:right+llen]

            if word in target:
                seen[word]+=1
                count+=1
                while seen[word]>target[word]:
                    left_word = s[left:left+llen]
                    seen[left_word]-=1
                    left+=llen
                    count -=1
                if count == len(words):
                    res.append(left)

                    # 得到结果以后将左边那个给退出来，继续处理
                    left_word = s[left:left+llen]
                    left+=llen
                    seen[left_word]-=1
                    count-=1
            else:
                seen.clear()
                count = 0
                left = right+llen
    return res
```

## 方法二，暴力解法

```python
def findSubstring(self, s: str, words: List[str]) -> List[int]:
    slen = len(s)
    llen = len(words[0])
    wlen = len(words)*llen
    if slen<wlen:
        return []
    target = Counter(words)   # 修改：记录 words 中每个单词需要出现的次数
    res = []

    for i in range(slen-wlen+1):

        start = i
        seen = Counter()
        while start<i+wlen:
            cur = s[start:start+llen]
            if cur not in target:
                break
            seen[cur]+=1
            
            if seen[cur]>target[cur]:
                break
            start+=llen
        if start == i +wlen:
            res.append(i)
        
    return res
        
```

# 题目**433**✨✨

```python
from typing import Deque
```

```python
def minMutation(self, startGene: str, endGene: str, bank: List[str]) -> int:
    q = Deque()
    if endGene not in bank:
        return -1
    q.append((startGene,0))
    bank = set(bank)
    while q:
        start,step = q.popleft()
        for i in range(len(start)):
            for y in ("A","C","G","T"):
                temp = start[:i]+y+start[i+1:]
                if temp not in bank:
                    continue
                if temp == endGene:
                    return step+1
                bank.remove(temp)
                q.append((temp,step+1))

    return -1
```
