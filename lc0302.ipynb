```python
from typing import List
```

>本题像是一个排序问题，找到合适的排序序列，然后用到变换中去  
>但是需要注意的是，如果当前是第i个序列，然后向下寻找到第一个符合条件的j就可以用来作为想要变换的目标值，对于j而言，如果下面还有符合条件的，那么j就是距离最近的，如果下面没有符合条件的，那么也得是要j。  
>如果是因为j被换掉以后，导致后面的行没有办法满足内容，那么只能说那本身就不符合内容要求了，因为满足了下面的要求，当前的值就没法满足了。从而体现出了贪心的策略

```python
def minSwaps(self, grid: List[List[int]]) -> int:
    m = len(grid)
    n = len(grid[0])
    one = [-1]*n
    res = 0
    # 确定好每一列中最后一个1出现的位置
    for i in range(m):
        for j in range(n-1,-1,-1):
            if grid[i][j] ==1:
                one[i] = j
                break

    for i in range(n):
        k = -1
        for j in range(i,n):
            if one[j]<=i:
                res+=j-i
                k = j
                break
        if k == -1:
            return -1
        for j in range(k,i,-1):
            one[j],one[j-1] = one[j-1],one[j]

    return res
```

# 题目**66**✨

```python
def plusOne(self, digits: List[int]) -> List[int]:
    digits.reverse()
    last = 1
    for i in range(len(digits)):
        temp = (digits[i]+last)%10
        last = (digits[i]+last)//10
        digits[i] = temp
    if last:
        digits.append(1)
    return digits.reverse()
```

```python
def plusOne(self, digits: List[int]) -> List[int]:
    last = 1
    for i in range(len(digits)-1,-1,-1):
        temp = (digits[i]+last)%10
        last = (digits[i]+last)//10
        digits[i] = temp
    if last:
        return [1]+digits
    return digits
```

# 题目**69**✨

```python
def mySqrt(self, x: int) -> int:
    for i in range(x):
        if i*i==x:
            return i
        if i*i>x:
            return i-1
```

## 方法二，用二分查找的方法

```python
def mySqrt(self, x: int) -> int:
    left,right,ans = 0,x,-1
    while left<=right:
        mid = (left+right)//2
        if mid * mid<=x:
            ans = mid
            left = mid+1
        else:
            right = mid-1
    return ans
```

```python
def myPow(self, x: float, n: int) -> float:
    if n == 0:
        return 1
    if n<0:
        n = -n
        x = 1/x
    ans = 1

    while n>0:
        i = 1
        ttt = 0
        while i*2<=n:
            i*=2
            ttt+=1
        part = x
        for _ in range(ttt):
            part *=part
        ans *=part
        n-=i
    return ans
    



    
    
    
```
