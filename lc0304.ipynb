# 题目**1582**✨✍️

```python
from typing import List
```

```python
def numSpecial(self, mat: List[List[int]]) -> int:
    m = len(mat)
    n = len(mat[0])
    rows = [0]*m
    cols = [0]*n
    for i in range(m):
        for j in range(n):
            rows[i]+=mat[i][j]
            cols[j]+=mat[i][j]
    res = 0
    for i in range(m):
        if rows[i]!=1:
            continue
        for j in range(n):
            if cols[j]!=1:
                continue
            if mat[i][j] ==1:
                res+=1
                break
    return res
```

# 题目**139**✨✨

```python
def wordBreak(self, s: str, wordDict: List[str]) -> bool:
    n = len(s)
    wordDict = set(wordDict)
    dp = [False]*(n+1)
    dp[0]=True
    for i in range(n+1):
        for j in range(i+1,n+1):
            if s[i:j] in wordDict:
                dp[j] = True
    return dp[n]
```

# 题目**322**✨✨

```python
def coinChange(self, coins: List[int], amount: int) -> int:
    coins.sort(reverse=True)
    dp = [0]*(amount+1)
    dp[0] = 0
    for i in range (amount+1):
        for j in coins:
            if j>i:
                break
            dp[i] = min(dp[i],dp[i-j]+1)
    return dp[amount]
```

# 题目**300**✨✨

```python
def lengthOfLIS(self, nums: List[int]) -> int:
    n = len(nums)
    dp = [1]*n

    for i in range(n):
        for j in range(i):
            if nums[j]<nums[i]:
                dp[i] = max(dp[i],dp[j]+1)
    return max(dp)
```

# 题目**120**✨✨

```python
def minimumTotal(self, triangle: List[List[int]]) -> int:
    n = len(triangle)
    dp = [inf]*((n*n+n)//2)

    dp[0]= triangle[0][0]
    for i in range(1,len(triangle)):
        for j in range(len(triangle[i])):
            if 0<j<i:
                dp[(i*i+i)//2+j] = min(dp[(i*i-i)//2+j],dp[(i*i-i)//2+j-1])+triangle[i][j]
            elif j==0:
                dp[(i*i+i)//2] = dp[(i*i-i)//2]+triangle[i][j]
            elif j ==i:
                dp[(i*i+i)//2+j] = dp[(i*i-i)//2+j-1]+triangle[i][j]
    t = len(triangle[-1])
    return min(dp[(t*t-t)//2:])


```
