# 题目**3070**✨✨

```python
from typing import List
```

```python
def countSubmatrices(self, grid: List[List[int]], k: int) -> int:
    res = 0
    m = len(grid)
    n = len(grid[0])

    for i in range(0,m):
        for j in range(0,n):
            if i ==0 and j ==0:
                if grid[0][0]<=k:
                    res+=1
                continue 
            if i == 0:
                grid[0][j] = grid[0][j]+grid[0][j-1]
            elif j == 0:
                grid[i][0] = grid[i][0]+grid[i-1][0]
            else:
                grid[i][j] = grid[i][j]+grid[i-1][j]+grid[i][j-1]-grid[i-1][j-1]
            if grid[i][j]<=k:
                res+=1
            else:
                break
        return res            
```

# 题目**3**✨✨

```python
def lengthOfLongestSubstring(self, s: str) -> int:
    res = 0
    hash = set()
    left ,right = 0,0
    while right<len(s):
        if s[right] not in hash:
            hash.add(s[right])
            res = max(res,right-left+1)
            right+=1
        else:
            while s[right] in hash:
                hash.remove(s[left])
                left+=1
    return res
```

# 题目**438**✨✨

```python
def findAnagrams(self, s: str, p: str) -> List[int]:
    slen,plen = len(s),len(p)
    if slen<plen:
        return []
    ans = []
    scount =[0]*26
    pcount =[0]*26
    for i in range(plen):
        scount[ord(s[i])-97]+=1
        pcount[ord(p[i])-97]+=1
    
    if pcount == scount:
        ans.append(0)

    for i in range(slen-plen):
        scount[ord(s[i])-97]-=1
        scount[ord(s[i+plen])-97]+=1
        if scount == pcount:
            ans.append(i+1)
    return ans

```
