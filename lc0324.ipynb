```python
from typing import List
```

```python
def constructProductMatrix(self, grid: List[List[int]]) -> List[List[int]]:
    m = len(grid)
    n = len(grid[0])
    back = [[1]*n for _ in range(m)]
    for i in range(m-1,-1,-1):
        for j in range(n-1,-1,-1):
            if i== m-1 and j == n-1: 
                continue
            if j == n-1:
                back[i][j] = back[i+1][0]*grid[i+1][0]
                continue
            back[i][j] = back[i][j+1]*grid[i][j+1]
    prefix = 1

    for i in range(m):
        for j in range(n):
            back[i][j] = prefix*back[i][j]
            prefix *=grid[i][j]
    
    return back
            
```

## 方法二 该版本会超时

```python
def constructProductMatrix(self, grid: List[List[int]]) -> List[List[int]]:
    counter_zero = 0
    multi = 1
    m = len(grid)
    n = len(grid[0])
    for i in range(m):
        for j in range(n):
            if grid[m][n] == 0:
                counter_zero+=1
            else:
                multi*=grid[m][n]

    if counter_zero>=2:
        return [[0]*n for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 0:
                grid[i][j] = multi
            else:
                if counter_zero ==1:
                    grid[i][j] = 0
                else:
                    grid[i][j] = multi//grid[i][j]
    return grid
```

# 题目**238**✨✨

```python
def productExceptSelf(self, nums: List[int]) -> List[int]:
    n = len(nums)
    res = [1]*n
    for i in range(n-2,-1,-1):
        res[i] = res[i+1]*nums[i+1]
    prefix = 1

    for i in range(n):
        res[i] = res[i]*prefix
        prefix=prefix*nums[i]

    return res
```

# 题目**78**✨✨

```python
def subsets(self, nums: List[int]) -> List[List[int]]:
    res = set()
    n = len(nums)
    curr = list()
    def dfs(t:int):
        res.add(tuple(curr))
        for i in range(t,n):
            curr.append(nums[i])
            dfs(i+1)
            curr.pop()
    dfs(0)
    return [list(i) for i in res]
```

# 题目**17**✨✨

```python
def letterCombinations(self, digits: str) -> List[str]:
    hash = {
        "2":['a','b','c'],
        "3":['d','e','f'],
        "4":['g','h','i'],
        "5":['j','k','l'],
        "6":['m','n','o'],
        "7":['p','q','r','s'],
        "8":['t','u','v'],
        "9":['w','x','y','z']
    }
    res = set()
    curr = []
    n = len(digits)
    def dfs(start:int):
        if start == n:
            res.add("".join(curr))
            return
        for i in range(start,n,1):
            for j in hash[digits[i]]:
                curr.append(j)
                dfs(i+1)
                curr.pop()
    dfs(0)
    return res
        
        
```

# 题目**70**✨

```python
def climbStairs(self, n: int) -> int:
    if n ==1:
        return 1
    if n ==2:
        return 2
    last = 2
    llast = 1
    for i in range(3,n):
        t= last
        last = last+llast
        llast = t
    return last+llast
    
```
