```python
from typing import List
```

```python
def minimumDifference(self, nums: List[int], k: int) -> int:
    if len(nums)<=1:
        return 0
    if k>len(nums):
        return 0
    nums.sort()
    res = float("inf")
    for i in range(k-1,len(nums)):
        if nums[i]-nums[i-k+1]<res:
            res = nums[i]-nums[i-k+1]
    return res
```

题目**51**

```python
def solveNQueens(self, n: int) -> List[List[str]]:
    loc = [["."]*n for _ in range(n)]
    res=[]
    col = [False]*n
    diag1 = [False]*(2*n)
    diag2 = [False]*(2*n)
    def dfs(row:int):
        if row == n:
            temp =["".join(r) for r in loc]
            res.append(temp)
            return
        for c in range(n):
            if col[c] or diag1[row-c+n] or diag2[row+c]:
                continue
            loc[row][c]="Q"
            col[c] = True
            diag1[row-c+n]=True
            diag2[row+c]= True
            dfs(row+1)
            loc[row][c]="."
            col[c] = False
            diag1[row-c+n]=False
            diag2[row+c]= False
    dfs(0)

            
    return res
                
                
```

```python
def largestRectangleArea(self, heights: List[int]) -> int:
    # 这个题目使用单调栈做
    # 先用三次遍历去做
    n = len(heights)

    st = []
    left = [-1]*n

    # 左边界
    for i,h in enumerate(heights):
        while st and heights[st[-1]]>h:
            st.pop()
        if st:
            left[i] = st[-1]
        st.append(i)

    # 右边界
    right = [n]*n
    st.clear()
    
    for i in range(n-1,-1,-1):

        h = heights[i]
        while st and heights[st[-1]]>=h:
            st.pop()
        if st:
            right[i] = st[-1]
        st.append(i)
    
    res = 0

    for h,l,r in zip(heights,left,right):
        res = max(res,h*(r-l-1))
    return res


        
```

```python
def largestRectangleArea(self, heights: List[int]) -> int:
    n = len(heights)
    left = [-1]*n
    right = [n]*n
    st=[]
    for i,h in enumerate(heights):
        while st and heights[st[-1]]>=h:
            # st.pop()
            right[st.pop()] = i
        if st:
            left[i] = st[-1]
        st.append(i)
    res = 0
    for h,l,r in zip(heights,left,right):
        res = max(res,h*(r-l-1))
    return res


```

```python
# 依旧是这个题目，但是是单调栈
def largestRectangleArea(self, heights: List[int]) -> int:
    heights.append(-1)
    st = [-1]
    ans = 0
    for right,h in enumerate(heights):
        while len(st)>1 and heights[st[-1]]>=h:
            i = st.pop()
            left = st[-1] #这个是刚才st[-1]（被pop了）的最大左边界
            # right对应的height[right]就是h，比i对应的height[i]高，是i对应的最近右边界
            ans = max(ans,heights[i]*(right-left-1))
        st.append(right)
    return ans
    
```
