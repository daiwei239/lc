````python
# 643. Maximum Average Subarray I
def findMaxAverge(self,nums:list[int],k:int) ->float:
    maxnum=0
    currsum=0
    for i in range(len(nums)):
        currsum+=nums[i]
        if(i<k-1):
            continue
        else:
            maxnum = max(maxnum,currsum)
        currsum-=nums[i-k+1]
    return maxnum/k
````

````python
#1456定长子串元音最大值
def maxVowels(self,s:str,k:int) -> int:
    curr = 0
    maxvo=0
    for i,c in enumerate(s):
        if c in "aeiou":
            curr+=1
        if i <k-1:
            continue
        maxvo = max(maxvo,curr)
        if s[i-k+1] in "aeiou":
            curr-=1
    return maxvo
        
````

````python
#34
from typing import List
def lower_bound(self,nums:List[int],target:int)->int:
    left = -1
    right = len(nums)
    while left+1<right:
        mid = left+(right-left)//2
        if nums[mid]<target:
            left = mid
        else:
            right = mid
    return right


def searchRange(self,nums: List[int],target:int) ->List[int]:
    start = self.lower_bound(nums,target)
    if start ==len(nums)or nums[start]!=target:
        return [-1,1]
    end = self.lower_bound(nums,target+1)
    return [start,end]

````

````python
#2529
# 这个题目的含义就是找到大于0的第一个整数的位置，
# 以及找到小于0的最大整数的位置
def position(self,nums:List[int],target:int)->int:
    left = -1
    right = len(nums)
    while left+1<right:
        mid=left+(right-left)//2
        if nums[mid]<=target:
            left = mid
        else:
            right=mid
    return right


def maximumCount(self, nums: List[int]) -> int:
    start = self.position(nums,0)
    if start==len(nums) or nums[start]!=0:
        return len(nums)
    end = self.position(nums,1)
    return max(start+1,len(nums)-end)
````

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
````

>题目92中，先要将其遍历到要进行处理的第一个节点的前一个节点

>对于使用了头节点的方式来说，其循环的次数也就只需要left-1次

>比如说当前要处理的left节点是3，

>第一次循环，p0=node1

>第二次循环 p0=node2

>所以只需要循环两次就可以了

>在中间需要调节的部分完成后，需要和整个链路上的大方向进行对齐

>将p0.next(原本链路上的头节点).next=curr(结束反转后的第一个主链路节点)

>p0.next = pre

````python
#92
def reverseBetween(self,head:Optional[ListNode],left:int,right:int)->Optional[ListNode]:
    dummy = ListNode(next=head)
    p0=dummy
    for _ in range(left-1):
        p0=p0.next
    
    pre=None
    curr=p0.next
    for _ in range (right-left+1):
        nxt=curr.next
        curr.next=pre
        pre = curr
        curr = nxt
    p0.next.next=curr
    p0.next = pre
    return dummy.next

````

````python
#25
def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    dummy = ListNode(next=head)
    n=0
    p0=dummy
    # 计算整个数组的长度
    while p0!=None:
        n+=1
        p0=p0.next
    p0=dummy
    pre=None
    curr = dummy.next
    while n>k:
        n-=k
        # 开始做链表的反转
        for _ in range(k):
            nxt = curr.next
            curr.next = pre
            pre = curr
            curr=nxt
        nxt=p0.next
        p0.next.next=curr
        p0.next=pre
        p0=nxt

    return dummy.next
````

````python
#141
def hasCycle(self,head:Optional[ListNode])->bool:
    slow,fast=head,head
    while fast and fast.next:
        fast = fast.next.next
        slow = slow.next
        if fast is slow:
            return True
        
    return False
````

````python
#142
def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
    slow,fast = head,head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        # 得到了，两者的交汇点
        if fast is slow:
            while head is not slow:
                head=head.next
                slow = slow.next
            return slow
    return None

    
````

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
````

````python
#104
def maxDepth(self,root:Optional[TreeNode])->int:
    if root is None:
        return 0;
    ldep = self.maxDepth(root.left)
    rdep = self.maxDepth(root.right)
    return max(ldep,rdep)+1
````

````python
# 100
def isSameTree(self,p:Optional[TreeNode],q:Optional[TreeNode])->bool:
    if p is None or q is None:
        return (p is q)
    return (p.val==q.val)&self.isSameTree(p.left,q.left)&self.isSameTree(p.right,q.right)
````

````python
#101
def isSymmetric(self, root: Optional[TreeNode]) -> bool:
    if root is None:
        return True
    return self.orTrue(root.left,root.right)
    
def orTrue(self,root1:Optional[TreeNode],root2:Optional[TreeNode])->bool:
    if root1 is None or root2 is None:
        return root1 is None
    return (root1.val == root2.val)&self.orTrue(root1.left,root2.right)&self.orTrue(root1.right,root2.left)
````

````python
#110
def isBalanced(self, root: Optional[TreeNode]) -> bool:
    def get_height(node:Optional[TreeNode])->int:
        if node is None:
            return 0
        lheight = self.get_height(node.left)
        if lheight ==-1:
            return -1
        rheight = self.get_height(node.right)
        if rheight ==-1 or abs(lheight-rheight)>1:
            return -1
        return max(lheight,rheight)+1
    return self.get_height(root)!=-1
````
