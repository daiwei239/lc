```python
# 先做一个前缀树，用来查找当前的source[j+1:i]的编号是多少
class Trie:
    def __init__(self):
        self.child =[None]*26
        self.id = -1

def add(node:Trie,word: str,index = list[int])->int:
    for ch in word:
        i = ord(ch)-ord("a")
        if node.child[i] is None:
            node.child[i]= Trie()
        node = node.child[i]
    # 这个地方的意思是，如果到了最后一个节点的时候，
    # 发现其是没有被标记过的新节点，就需要对其进行标记
    # 否则就不需要进行标记，直接返回就可以了
    if node.id == -1:
        index[0]+=1
        node.id = index[0]
    return node.id

def update(x:int,y:int):
    if x ==-1 or y<x:
        return y
    return x

class Solution:
     def minimumCost(self, source: str, target: str, original: List[str], changed: List[str], cost: List[int]) -> int:
        n,m = len(source),len(original)
        root = Trie()
        # 构建floyd算法的图
        # 构建节点的编号
        p= [-1]#用list的话就可以改变对应的值，成为全局变量
        G = [[inf]*(2*m) for _ in range(2*m)]

        # 初始化G
        for i in range(2*m):
            G[i][i] =0
        # 开始对source和target的具体值进行编号

        # source和target数组的大小是相同的
        for i in range(m):
            x = add(root,source[i],p)
            y = add(root,target[i],p)
            G[x][y] = min(G[x][y],cost[i])#初始化floyd图的值

        # 更新floyd图的最终值
        for k in range(p[0]+1):
            for i in range(p[0]+1):
                for j in range(p[0]+1):
                    G[i][j] = min(G[i][j],G[i][k]+G[k][j])

        # floyd的部分结束，然后进行dp的部分

        f = [-1]*n
        # 更新最后的结果，用动态规划
        # 从0开始遍历
        for j in range(n):
            # 这一段说明当前没有合适的匹配的值，直接进行下一个循环即可
            if j>0 and f[j-1]==-1:
                continue
            
            base =0 if j==0 else f[j-1]#base是前一项的最小cost值

            # 开始更新值
            # 1.当前的source[j]==target[j]
            if source[j]==target[j]:
                f[j] = update(f[j],base)
            # 2.当前的source[j]!= target[j]

            u,v = root

            for i in range(j,n):
                # 对source[j:i],target[j:i]是否在original,changed 数组中进行判断
                # 这个检索的过程是用root遍历的，
                # 全部的original和changed的数组都在root下的前缀树上
                u = u.child[ord(source[i])-ord("a")]
                v = v.child[ord(target[i])-ord("a")]
                
                if u is None or v is None:
                    break
                if u.id != -1 and v.id != -1 and G[u.id][v.id]!=inf:
                    f[i] = update(f[i],base+G[u.id][v.id])
        
        return f[n-1]
```

# 题目**189**✨✨

```python
def rotate(self, nums: List[int], k: int) -> None:
    n = len(nums)
    times = n-k
    for i in range(times):
        nums.append(nums[i])
    nums = nums[times+1:]
```

>采用三次反转操作

```python
def rotate(self, nums: List[int], k: int) -> None:
    def reverse(i:int ,j:int):
        while i<j:
            nums[i],nums[j]=nums[j],nums[i]
            i+=1
            j-=1
    n = len(nums)
    reverse(0,n-1)
    reverse(0,k)
    reverse(k+1,n-1)
```

# 题目**121**✨

```python
def maxProfit(self, prices: List[int]) -> int:
    minprice = float("inf")
    maxprofit = 0
    for i in range(len(prices)):
        minprice = min(minprice,prices[i])
        maxprofit = max(maxprofit,prices[i]-minprice)
    return maxprofit
```
