# 题目**3643**✨

```python
from typing import List
```

```python
def reverseSubmatrix(self, grid: List[List[int]], x: int, y: int, k: int) -> List[List[int]]:
    up_point = x
    down_point = x+k
    while up_point<down_point:
        grid[up_point][y:y+k+1],grid[down_point][y:y+k+1]=\
        grid[down_point][y:y+k+1],grid[up_point][y:y+k+1]
        up_point+=1
        down_point-=1
    return grid
```

# 题目**226**✨

```python
from typing import Optional
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
    def dfs (root:Optional[TreeNode])->Optional[TreeNode]:
        if root is None:
            return None
        leftNode =dfs(root.left)
        rightNode = dfs(root.right)
        root.left = rightNode
        root.right = leftNode
        return root

    return dfs(root)
```

# 题目**101**✨

```python
def isSymmetric(self, root: Optional[TreeNode]) -> bool:
    def dfs(r1:Optional[TreeNode],r2:Optional[TreeNode])->bool:
        if r1 and not r2:
            return False
        if not r1 and r2:
            return False
        if r1.val != r2.val:
            return False
        if not r1 and not r2:
            return True
        l  = dfs(r1.left,r2.right)
        r = dfs(r1.right,r2.left)
        return l and r
    if root is None:
        return True
    return dfs(root.left,root.right)
```

# 题目**543**✨

```python
def diameterOfBinaryTree(self, root: Optional[TreeNode]) -> int:
    maxlen = 0
    def dfs(root:Optional[TreeNode])->int:
        nonlocal maxlen
        if root is None:
            return 0
        left = dfs(root.left)
        right = dfs(root.right)
        maxlen = max(maxlen,left+right)
        return max(left,right)+1
    dfs(root)
    return maxlen
```

# 题目**108**✨

```python
def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
    def dfs(left:int,right:int)-> Optional[TreeNode]:
        if left>right:
            return None
        mid = left+(right-left)//2
        
        left = dfs(left,mid-1)
        right = dfs(mid+1,right)
        return TreeNode(nums[mid],left,right)
    
    left,right = 0,len(nums)-1
    return dfs(left,right)
```

# 题目**102**✨✨

```python
from collections import deque
```

```python
def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
    if root is None:
        return []
    q = deque()
    q.append(root)
    res = []
    res.append([root.val])
    
    while q:
        temp = []
        for _ in range(len(q)):
            node=q.popleft()        
            if node.left:
                temp.append(node.left)
            if node.right:
                temp.append(node.right)
        if temp!=None:
            res.append([t.val for t in temp])
            q.extend(temp)
    return res
        
```

# 题目**98**✨✨

```python
def isValidBST(self, root: Optional[TreeNode]) -> bool:
    last = -inf
    def dfs(root:Optional[TreeNode])->bool:
        nonlocal last
        if root is None:
            return True
        le=dfs(root.left)
        if root.val <=last:
            return False
        last = root.val
        ri = dfs(root.right)
        return le and ri
    return dfs(root)
```

# 题目**230**✨✨

```python
def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
    count = 0
    res = 0
    def dfs(root:Optional[TreeNode]):
        nonlocal count,res
        if root is None:
            return
        dfs(root.left)
        count+=1
        if count == k:
            res = root.val
            return
        dfs(root.right)
    dfs(root)
    return res
```

# 题目**199**✨✨

```python
def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
    if not root:
        return []
    res = []
    q = deque()
    q.append(root)
    while q:
        level_size = len(q)
        for i in range(level_size):
            node = q.popleft()
            if node.left:
                q.append(node.left)
            if node.right:
                q.append(node.right)
            if i == level_size-1:
                res.append(node.val)
        
    return res
    
```
