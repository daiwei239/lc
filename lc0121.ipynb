```python
def partitionLabels(self, s: str) -> List[int]:
    # 对于问题，先预处理，将整个s的全部字母出现的最大位置给记录下来，
    # 然后填充到和s相对应的位置上
    # 然后使用贪心算法，给当前的全部内容进行一个坐标的固定
    pos = dict()
    for i,alp in enumerate(s):
        pos[alp] = i
    
    loc = list()
    for i in range(s):
        loc.append(pos.get(s[i]))
    
    res = []
    

    # 不需要将这些元素给保存下来
    for  i in range(len(s)):
        j = loc[i]
        left =i
        while i <=j:
            if loc[i]>j:
                j=loc[i]
            i+=1
        res.append(j-left+1)
    return res
```

题目**300**
>采用这种dp的方式而言，比较直接
>就是看当前状态下，是保持原本的dp合适还是在dp[j]的基础上加1更好
>同时需要注意，每个i都会将原本的值全部给遍历一遍，并且不一定就是结尾的值就是最大的

```python
def lengthOfLIS(self, nums: List[int]) -> int:
    # 申请一个dp数组，然后对于以i为终点的值，每次运行的时候都进行刷新
    n = len(nums)
    dp = [1 for _ in range(n)]
    for i in range(n):
        for j in range(i):
            if nums[i]>nums[j]:
                dp[i] = max(dp[i],dp[j]+1)
    return max(dp)
```

```python
def findDuplicate(self, nums: List[int]) -> int:
    slow,fast =0,0
    while True:
        slow=nums[slow]
        fast =nums[fast]
        fast = nums[fast]
        if fast == slow:
            break
    number =0
    while number!=slow:
        number= nums[number]
        slow = nums[slow]
    return slow
    
```

```python
#采用二分查找的方式去做
# 虽然采用了二分查找的思想，但是不是完全按照二分查找去做的，
# 在关于个数的统计上都是不一样的
def findDuplicate(self, nums: List[int]) -> int:
    minN,maxN = 1,len(nums)-1
    while minN<maxN:
        mid = (minN+maxN)/2
        total = 0
        for j in nums:
            if j<=mid:
                total+=1
        if total<maxN:
            minN=mid+1
        else:
            maxN = mid
    return minN 
```

```python
def minBitwiseArray(self, nums: List[int]) -> List[int]:
    for i in range(len(nums)):
        res =-1
        d=1
        while (nums[i] & d)!=0:
            res = nums[i]-d
            d<<=1
        nums[i]=res
    return nums
```

```python
def nextPermutation(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
    # 先从右往左遍历，找到第一个非升序排列的数字A
    # 然后从右往左遍历，找到第一个大于这个数字A的数字B
    # 二者交换
    # 交换后，将更新后的B之后的数字进行升序排列即可
    # 因为原本A插入其中后是从左到右降序的，reversed一下即可变为升序排列
        n=len(nums)
        loc1=0
        loc2=0
        for i in range(n-1,-1,-1):
                if(nums[i]>nums[i-1]):
                        loc1=i-1
                        break
        for j in range(n-1,-1,-1):
                if nums[j]>nums[loc1]:
                        loc2=j
                        break
                
        temp = nums[loc1]       
        nums[loc1]=nums[loc2]
        nums[loc2]=temp

        part=nums[loc1+1:]
        part.reverse()
        for i in range(loc1+1,n):
                nums[i]=part[i-loc1-1]         
```

题目**416**

```python
def canPartition(self, nums: List[int]) -> bool:
    # 对于本题，先计算出全部元素的总和
    # 然后再对元素进行排序
    # 使用动态规划
    total =0
    for i in nums:
        total+=i
    if total % 2 != 0:
        return False
    total = total // 2
    dp = [False]*(total+1)
    dp[0]=True
    
    for x in nums:
        for i in range(total,x-1,-1):
            dp[i]=dp[i] or dp[i-x]

    return dp[total] 
```

```python
from typing import Optional
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def maxPathSum(self, root: Optional[TreeNode]) -> int:
        # 这一题所采用的方式是
        # 单独加一个全局变量，然后使用dfs深度搜索去比较返回出最大的
        # 路径和
        ans = -inf
        def dfs(root:Optional[TreeNode])->int:
            if root is None:
                return 0
            lval = dfs(root.left)
            rval = dfs(root.right)
            ans = max(ans,lval+rval+root.val)
            return max(max(lval,rval)+root.val,0)
        
        return ans


```
