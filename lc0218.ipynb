# 题目**693**✨

```python
def hasAlternatingBits(self, n: int) -> bool:
    s=bin(n)
    if len(s)<2:
        return True
    left,right =0,1
    while right<len(s):
        if s[left] ==s[right]:
            return False
        left += 1
        right += 1
    return True
```

# 题目**108**✨✨

```python
# Definition for a Node.
class Node:
    def __init__(self, val = 0, neighbors = None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []
```

```python
from typing import Optional, List
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python

def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
    def build(left:int,right:int):
        if left>right:
            return None
        mid =(left+right)//2
        res = Node(nums[mid])
        res.left = build(left,mid-1)
        res.right = build(mid+1,right)
        return res 
    return build(0,len(nums)-1)

        
```

# 题目**133**✨✨

```python
def cloneGraph(self, node: Optional['Node']) -> Optional['Node']:
    visited= dict()
    def dfs(node:Optional['Node'])->Optional['Node']:
        if node is None:
            return None
        
        res = Node(node.val)
        if node not in visited:
            visited[node]= res

        for nod in node.neighbors:
            if nod not in visited:
                t = dfs(nod)
                res.neighbors.append(t)
            else:
                res.neighbors.append(visited[nod])
        return res
    return dfs(node)
            
        
```
