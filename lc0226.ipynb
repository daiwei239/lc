# 题目**1404**✨✨

```python
def numSteps(self, s: str) -> int:
    counter = 0
    dec = int(s,2)
    while dec != 1:
        if dec %2 !=0:
            counter+=1
            dec+=1
        
        dec = dec//2
        counter+=1
        
    return counter
```

## 方法二，用字符串的方式

```python
def numSteps(self, s: str) -> int:
    res = 0
    carry = 0
    for i in range(len(s)-1,-1,-1):
        if carry+int(s[i]) == 1:
            res+=2
            carry = 1
        else:
            res +=1
    return res+carry
```

# 题目**162**✨✨

```python
def findPeakElement(self, nums: List[int]) -> int:
    if len(nums)<2:
        return 0
    def get_max(left:int,right:int)->int:
        if left  = 0 and nums[left]>nums[left+1]:
            return nums[left]
        if right = len(nums)-1 and nums[right]>nums[right-1]:
            return nums[right]
        mid = (left+right)//2
        if nums[mid]>nums[mid+1] and nums[mid]>nums[mid-1]:
            return nums[mid]
        elif nums[mid+1]>nums[mid]:
            return get_max(mid,right)
        else:
            return get_max(left,mid)
    return get_max(0,len(nums)-1)
```

# 题目**33**✨✨

```python
from typing import List
```

```python
def search(self, nums: List[int], target: int) -> int:
    left,right = 0,len(nums)-1
    while left<right:
        mid = (left+right)//2
        if nums[mid] == target:
            return mid
        if nums[left]<nums[mid]:
            if target<nums[mid] and target>=nums[left]:
                right = mid-1
            else:
                left = mid+1
        else:
            if nums[mid]<target and target<=nums[right]:
                left = mid+1
            else:
                right = mid-1
    return -1
```
