```python
from typing import List
from itertools import pairwise
from collections import defaultdict
from heapq import heapify, heappop, heappush
```

```python
def minimumPairRemoval(self, nums: List[int]) -> int:
    # 对于这一题，好像没有什么思路去实现
    # 本题的核心就是，每次筛选出当前和最小的两个相邻值
    # 将其合并，然后一直循环直到其成为一个非递减的数组
    # 这个题目和昨天的相比主要是采用了一些高级的数据结构，
    # 降低了时间复杂度
    n = len(nums)
    h = [] #存放相邻元素的和以及左边元素的下标
    dec = 0 #记录当前不是递增的元素对的个数

    for i,(x,y) in enumerate(pairwise(nums)):
        if x>y:
            dec+=1
        h.append((x+y,i))#这一步实现了存储各相邻元素的和以及其左边元素的下标
    
    heapify(h)#创建出了一个最小堆，用来每次得出，最小的元素和
    lazy = defaultdict(int)
    # 这个是每个元素i的左右元素的下标
    left = [i for i in range(-1,n)]
    right = [i for i in range(1,n+1)]

    ans = 0 #运算的次数
    while dec:
        ans +=1 #加一个操作次数
        # 获取最小的下标值
        while lazy[h[0]]:#如果有lazy值，就说明了当前的值是无效的值，弹出来就可以了
            lazy[heappop(h)]-=1
        s,i = heappop(h)#要给它弹出来，左元素的下标是i

        # 获取右元素
        nex = right[i]
        if nums[i]>nums[nex]:
            dec-=1

        # 当前元素的下标和前面一个元素形成的数字已经不能用了
        pre = left[i]
        if pre>=0:
            if nums[pre]>nums[i]:
                dec-=1
            if nums[pre]>s:
                dec+=1
            
            # 更新堆和lazy区域
            lazy[(nums[pre]+nums[i],pre)]+=1
            heappush(h,(nums[pre]+s,pre))
        
        # 当前原左元素的下个下标（原右元素）的下一个坐标的内容
        nex2 = right[nex]
        if nex2<n:
            if nums[nex]>nums[nex2]:
                dec-=1
            if s > nums[nex2]:
                dec+=1
            # 更新堆和lazy部分
            lazy[(nums[nex]+nums[nex2],nex)]+=1
            # heappush(h,(s+nums[nex2],nex))
            heappush(h,(s+nums[nex2],i))
        nums[i] = s
        # 删除nex元素
        l,r = left[nex],right[nex]
        right[l] = r
        left[r] = l
    return ans
```

```python
from typing import Optional
```

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    if not head  or not head.next:
        return head
    pre = None
    nxt =head.next
    while head and nxt:
        head.next = pre
        pre = head
        head=nxt
        nxt = nxt.next
    head.next = pre
    return head
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
class Solution:
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        # 这道题感觉和206题目很像，只是多了一个参数的样子
        # 测算单个部分的翻转情况应该是如何的
        def reverseNode(head:Optional[ListNode],pre:Optional[ListNode],k:int):
            newTail = head
            nxt = head.next
            i =0
            while head and nxt and i<k-1:
                head.next = pre
                pre = head
                head = nxt
                nxt = nxt.next
            head.next = pre
            return head,newTail
        
        curr = 0
        p = head
        while p:
            curr+=1
            p = p.next
        dummy = ListNode(0)
        dummy.next = head
        preTail = dummy

        while head and curr>=k:
            kth = head
            for _ in range(k-1):
                kth = kth.next
            nxt = kth.next
            newHead,newTail = reverseNode(head,nxt,k)

            preTail.next = newHead
            preTail = newTail
            head = nxt
            curr-=k
        return dummy.next
```
