# 题目**868**✨

```python
def binaryGap(self, n: int) -> int:
    t=bin(n)
    left = 0
    res = 0
    while left<len(t):
        if t[left]=="1":
            break
        left+=1
    if left ==len(t):
        return 0
    for right in range(left+1,len(t)):
        if right =="1":
            res = max(res,right-left)
            left = right
    return res

```

# 题目**208**✨✨

```python
from typing import List
```

```python
class Trie:

    def __init__(self):
        self.children = [None]*26
        self.isEnd = False
        

    def insert(self, word: str) -> None:
        node = self
        for i in word:
            index = ord(i)-ord("a")
            if not node.children[index]:
                node.children[index] = Trie()
            node = node.children[index]
        node.isEnd = True


        
    def searchPrefix(self,prefix:str)->"Trie":
        node = self
        for i in prefix:
            index = ord(i)-ord("a")
            if  not node.children[index]:
                return None
            node = node.children[index]
        return node


        
                    
    def search(self, word: str) -> bool:
        res = self.searchPrefix(word)
        if res:
            return res.isEnd == True
        return False

        

    def startsWith(self, prefix: str) -> bool:
        return self.searchPrefix(prefix) is not None
        
    
```

# 题目**17**✨✨

```python
def letterCombinations(self, digits: str) -> List[str]:
    res = []
    phones={
        '2':['a','b','c'],
        '3':['d','e','f'],
        '4':['g','h','i'],
        '5':['j','k','l'],
        '6':['m','n','o'],
        '7':['p','q','r','s'],
        '8':['t','u','v'],
        '9':['w','x','y','z']
    }
    def getRes(digits:str,curr:str):
        if len(digits) == 0:
            res.append(curr)
            return
        for i in phones[digits[0]]:
            curr +=i
            getRes(digits[1:],curr)
            curr = curr[:-1]
    getRes(digits,"")
    return res

        
        
```
