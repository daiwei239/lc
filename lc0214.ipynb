# 题目**799**✨✨

```python
def champagneTower(self, poured: int, query_row: int, query_glass: int) -> float:
    #利用类似杨辉三角的方式
    dp=[[0.0]*(i+1) for i in range(query_row+1)]
    dp[0][0]= poured

    for i in range(query_row):
        for j in range(len(dp[i])):
            overflow = max(0,dp[i][j]-1)
            dp[i+1][j] =overflow/2
            dp[i+1][j+1] = overflow/2
    return min(1,dp[query_row][query_glass])
```

# 题目**230**✨✨

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
from typing import Optional
```

```python
def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
    list = []
    def inorder(root:Optional[TreeNode]):
        if root is None:
            return 
        inorder(root.left)
        list.append(root.val)
        inorder(root.right)

    inorder(root)
    return list[k-1]

```

# 题目**98**✨✨

```python
def isValidBST(self, root: Optional[TreeNode]) -> bool:
    def inorder(root:Optional[TreeNode]):
        if root is None:
            return float("-inf")
        if root.left and root.left.val>root.val:
            return False
        if root.right and root.right.val<root.val:
            return False
        return inorder(root.left) and inorder(root.right)
    return inorder(root)
```

```python
def isValidBST(self, root: Optional[TreeNode]) -> bool:
    def judge(root:Optional[TreeNode],left:int,right:int)->bool:
        if root is None:
            return True
        v = root.val
        if root.val>=right or root.val <=left:
            return False
        if not judge(root.left,left,v):
            return False
        if not judge(root.right,v,right):
            return False
        return True
    return judge(root,float("-inf"),float("inf"))
```

# 题目**200**✨✨

```python
from typing import List
from collections import deque
```

```python
def numIslands(self, grid: List[List[str]]) -> int:
    col = len(grid)
    row = len(grid[0])
    totoal_num = 0
    de = deque()
    for i in range(col):
        for j in range(row):
            if grid[i][j] == "1":
                totoal_num+=1
                de.append((i,j))
                grid[i][j] = "0"
                while de:
                    x,y = de.popleft()
                    for a,b in [(1,0),(-1,0),(0,1),(0,-1)]:
                        if 0<=x+a<col and 0<=y+b<row and grid[x+a][y+b] == "1":
                            de.append((x+a,y+b))
                            grid[x+a][y+b]="0"
    return totoal_num

```
