# 题目**3713**✨✨

```python
from collections import defaultdict

def longestBalanced(self, s: str) -> int:
    n = len(s)
    res = 0
    for i in range(n):
        cnt = defaultdict(int)
        for j in range(i,n):
            cnt[s[j]]+=1
            if len(set(cnt.values()))==1:
                res = max(res,j-i+1)
    return res
```

# 题目**106**✨✨

```python
from typing import List,Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def buildTree(self, inorder: List[int], postorder: List[int]) -> Optional[TreeNode]:
    in_map = {val: i for i,val in enumerate(inorder)}
    def build(inst:int,inend:int,postst:int,postend:int)->Optional[TreeNode]:
        if inst>inend or postst>postend:
            return None
        rval = postorder[postend]
        root = TreeNode(rval)
        i = in_map[rval]
        
        # 这时得到中间节点的位置，
        leftlen = i-inst
        root.left = build(inst,inst+leftlen-1,postst,postst+leftlen-1)
        root.right = build(i+1,inend,postst+leftlen,postend-1)
        return root
    return build(0,len(inorder)-1,0,len(postorder)-1)
```

# 题目**117**✨✨

```python
# Definition for a Node.
class Node:
    def __init__(self, val: int = 0, left: 'Node' = None, right: 'Node' = None, next: 'Node' = None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next
```

```python
from collections import deque
```

```python
def connect(self, root: 'Node') -> 'Node':
    cnt = deque()
    res = root
    cnt.append(root)
    length = len(cnt)
    while cnt:
        counter = 0
        length = len(cnt)
        while cnt and  counter<length:
            temp = cnt.popleft()
            if length-counter>1:
                temp.next = cnt[0]
            if temp and temp.left:
                cnt.append(temp.left)
            if temp and temp.right:
                cnt.append(temp.right)
            counter+=1
    return res
```

# 题目**112**✨

```python
def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
    def dfs(root:Optional[TreeNode],targetSum: int)->bool:
        if not root:
            return False
        if root.val ==targetSum and not root.left and not root.right:
            return True
        left = dfs(root.left,targetSum-root.val)
        right = dfs(root.right,targetSum-root.val)
        return left or right
    return dfs(root,targetSum)
```

# 题目**129**✨✨

```python
def sumNumbers(self, root: Optional[TreeNode]) -> int:
    res = 0
    def dfs(root:Optional[TreeNode],curr):
        nonlocal res
        if not root:
            return 0
        if root and not  root.left and not root.right:
            res+= root.val+curr*10  
            return   
        temp = root.val+curr*10
        dfs(root.left,temp)
        dfs(root.right,temp)
        
    dfs(root,0)
    return res
    
```
