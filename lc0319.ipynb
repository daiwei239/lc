# 题目**3212**✨✨

>采用类似前缀和的方式，用两个单独的矩阵，记录当前的x，y的个数，然后使用遍历的方式，判断当前的x和y元素的个数是否一致

```python
from typing import List
```

```python
def numberOfSubmatrices(self, grid: List[List[str]]) -> int:
        m = len(grid)
        n = len(grid[0])
        x_count = [[0]*n for _ in range(m)]
        y_count = [[0]*n for _ in range(m)]
        def judge(ch:str,i:int,j:int):
            nonlocal x_count
            nonlocal y_count
            if i==0:
                x_count[i][j] = x_count[i][j-1]
                y_count[i][j] = y_count[i][j-1]
                if ch =="X":
                    x_count[i][j] +=1
                elif ch =="Y":
                    y_count[i][j] +=1
                return
            if j == 0:
                x_count[i][j] = x_count[i-1][j]
                y_count[i][j] = y_count[i-1][j]
                if ch =="X":
                    x_count[i][j]+=1
                elif ch =="Y":
                    y_count[i][j] +=1
                return

            x_count[i][j] = x_count[i-1][j]+x_count[i][j-1]-x_count[i-1][j-1]
            y_count[i][j] = y_count[i-1][j]+y_count[i][j-1]-y_count[i-1][j-1]
            if ch =="X":
                x_count[i][j]+=1
            elif ch == "Y":
                y_count[i][j] +=1

        
        if grid[0][0]=="X":
            x_count[0][0]=1
        elif grid[0][0] =="Y":
            y_count[0][0]=1
        res = 0
        for i in range(m):
            for j in range(n):
                if i==0 and j ==0:
                    continue
                judge(grid[i][j],i,j)
                if x_count[i][j]==y_count[i][j] and x_count[i][j]>0:
                    res +=1
        return res
```

## 方法二，更统一的写法

```python
def numberOfSubmatrices(self, grid: List[List[str]]) -> int:
        m = len(grid)
        n = len(grid[0])
        x_count = [[0]*n for _ in range(m)]
        y_count = [[0]*n for _ in range(m)]
        res = 0 

        for i in range(m):
            for j in range(n):
                x_count[i][j] = (x_count[i-1][j] if i>0 else 0)+(x_count[i][j-1] if j>0 else 0)-(x_count[i-1][j-1] if i>0 and j>0 else 0)
                y_count[i][j] = (y_count[i-1][j] if i>0 else 0)+(y_count[i][j-1] if j>0 else 0)-(y_count[i-1][j-1] if i>0 and j>0 else 0)
                if grid[i][j] =="X":
                    x_count[i][j]+=1
                elif grid[i][j] == "Y":
                    y_count[i][j] +=1
                    
                if x_count[i][j]==y_count[i][j] and x_count[i][j]>0:
                    res +=1
        return res
```

# 题目**53**✨✨

```python
def maxSubArray(self, nums: List[int]) -> int:
    if len(nums)<2:
        return nums[0]
    minVal = min(0,nums[0])
    res = nums[0]
    
    for i in range(1,len(nums)):
        nums[i] = nums[i]+nums[i-1]
        res = max(nums[i]-minVal,res)
        minVal = min(minVal,nums[i])
    return res
```

# 题目**56**✨✨

```python
def merge(self, intervals: List[List[int]]) -> List[List[int]]:
    intervals.sort(lambda x:x[0])
    res = []
    currLeft = intervals[0][0]
    currRight = intervals[0][1]
    for i,interval in enumerate(intervals):
        if interval[0]<=currRight:
            currRight = max(currRight,interval[1])
            continue
        else:
            res.append([currLeft,currRight])
            currLeft = interval[0]
            currRight= interval[1]
    return res


```

# 题目**189**✨✨

```python
def rotate(self, nums: List[int], k: int) -> None:
    def reverse(left:int,right:int):
        while left<right:
            nums[left],nums[right] = nums[right],nums[left]
            left+=1
            right-=1
    k = k%len(nums)    
    reverse(0,len(nums)-1)
    reverse(0,k-1)
    reverse(k,len(nums)-1)
```

## 方式二，直接切分

```python
def rotate(self, nums: List[int], k: int) -> None:
    n = len(nums)
    k %= n
    nums[:] = nums[n-k:] + nums[:n-k]
```

# 题目**238**✨✨

```python
def productExceptSelf(self, nums: List[int]) -> List[int]:
    n = len(nums)
    leftM =[1]*n
    rightM = [1]*n
    if n<2:
        nums.reverse()
        return nums
    for i in range(1,n):
        leftM[i] = leftM[i-1]*nums[i-1]
    for j in range(n-2,-1,-1):
        rightM[j] = rightM[j+1]*nums[j+1]
    
    res = []
    for i in range(n):
        res.append(leftM[i]*rightM[i])
    return res
```

# 题目**73**✨✨

```python
def setZeroes(self, matrix: List[List[int]]) -> None:
    m = len(matrix)
    n = len(matrix[0])
    rows =set()
    cols = set()
    for i in range(m):
        for j in range(n):
            if matrix[i][j] ==0:
                rows.add(i)
                cols.add(j)
    for i in range(m):
        for j in range(n):
            if i in rows or j in cols:
                matrix[i][j] = 0
```

# 题目**41**✨✨✨

```python
def firstMissingPositive(self, nums: List[int]) -> int:
    n = len(nums)
    for i in range(n):
        if nums[i]<0:
            nums[i] = n
    for i in range(n):
        nums[abs(nums[i])] = -abs(nums[abs(nums[i])])
    
    for i in range(n):
        if nums[i]>0:
            return i
    return n
```

# 题目**54**✨✨

```python
def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
    m = len(matrix)
    n = len(matrix[0])
    res = []
    dir = ((1,0),(0,1),(-1,0),(0,-1))
    i,j = 0,0
    curr = 0
    for _ in range(m*n):
        res.append(matrix[i][j])

        matrix[i][j] = None

        x = i + dir[curr][0]
        y = j + dir[curr][1]

        if x<0 or x>=m or y<0 or y>=n or matrix[x][y]==None:
            curr = (curr+1)%4
        i = i + dir[curr][0]
        j = j + dir[curr][1]
    return res
        
```

# 题目**160**✨

```python
from typing import List
from collections import Optional

class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
```

```python
def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> Optional[ListNode]:
    hash = set()
    while headA:
        if headA not in hash:
            hash.add(headA)
            headA = headA.next
    while headB:
        if headB in hash:
            return headB
        headB = headB.next
    return None
```

# 题目**206**✨

```python
def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
    last = None
    curr = head
    while curr:
        nxt = curr.next
        curr.next = last
        last = curr
        curr = nxt
    return last        
```

# 题目**234**✨

```python
from collections import deque
```

```python
def isPalindrome(self, head: Optional[ListNode]) -> bool:
    stack = deque()
    curr = head
    while curr:
        stack.append(curr.val)
        curr = curr.next
    curr = head
    while curr:
        if curr.val != stack.pop():
            return False
        curr = curr.next()
    return True
    
```

# 题目**141**✨

```python
def hasCycle(self, head: Optional[ListNode]) -> bool:
    hash = set()
    while head:
        if head not in hash:
            hash.add(head)
            head = head.next
        else:
            return True
    return False
        
```

## 方法二

```python
def hasCycle(self, head: Optional[ListNode]) -> bool:
    if not head or not head.next:
        return False
    slow = head
    fast = head.next
    while slow != fast:
        if not fast or not fast.next:
            return False
        slow = slow.next
        fast = fast.next.next
    return True
```
