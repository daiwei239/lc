```python
def constructTransformedArray(self, nums: List[int]) -> List[int]:
    # 直接开始进行循环即可
    res =list()
    step =0
    for i in range(len(nums)):
        if nums[i]>0:
            step = (i+nums[i])%len(nums)
        else:
            step= i-abs(nums[i]) if i-abs(nums[i])>0 else abs( (i-abs(nums[i]))%len(nums))
        res.append(nums[step])
    return res
```

# 题目**290**✨

```python
def wordPattern(self, pattern: str, s: str) -> bool:
    s=s.split()
    h1 = dict()
    h2 = dict()
    if len(pattern)!= len(s):
        return False
    for i,j in zip (pattern,s):
        if i not in h1 and j not in h2:
            h1[i] = j
            h2[j] = i
        else:
            if h1.get(i)!=j or h2.get(j)!= i:
                return False
    return True
```

# 题目**242**✨

```python
def isAnagram(self, s: str, t: str) -> bool:
    h1 = dict()
    h2 = dict()
    if len(s)!= len(t):
        return False
    for i,j in zip(s,t):
        h1[i] = h1.get(i,0)+1
        h2[j] = h2.get(j,0)+1
    
    for k,v in h1.items():
        if h2.get(k,0) != v:
            return False
    return True

        
```

# 题目**49**✨✨

```python
import collections
from typing import List
```

```python
def groupAnagrams(strs: List[str]) -> List[List[str]]:
    hash = collections.defaultdict(list)
    for st in strs:
        k = "".join(sorted(st))
        hash[k].append(st)
    return list(hash.values())
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(groupAnagrams(strs))
```

```python
def twoSum(self, nums: List[int], target: int) -> List[int]:
    ha = dict()

    for i in range(len(nums)):
        ha[nums[i]] = i
        if target-nums[i] in ha:
            return [ha[target-nums[i]],i]
    
    return [0,0]
```

# 题目**202**✨

```python
def isHappy(self, n: int) -> bool:
    def getNext(curr:int)->int:
        total = 0
        while curr>0:
            total+=(curr%10)**2
            curr=curr/10
        return total
    
    ha = set()
    while n!= 1 and n not in ha:
        ha.add(n)
        n = getNext(n)
    
    return n ==1
```

# 题目**219**✨

```python
def containsNearbyDuplicate(self, nums: List[int], k: int) -> bool:
    ha = dict()
    for i in range(len(nums)):
        if nums[i] not in ha:
            ha[nums[i]] = i
        else:
            if abs(i-ha[nums[i]])<=k:
                return True
    return False
```

# 题目**128**✨✨

```python
def longestConsecutive(self, nums: List[int]) -> int:
    ha = dict()
    nums.sort()
    for i in range(len(nums)):
        ha[nums[i]] = ha.get(nums[i]-1)+1
    return max(ha.values())
```

## 方法2

>更快的一种写法方式

```python
def longestConsecutive(self, nums: List[int]) -> int:
    nums = set(nums)
    res = 0
    for i in nums:
        if i-1 not in nums:
            curr = i
            len = 1
            while curr+1 in nums:
                curr+=1
                len+=1
            res = max(res,len)
    return res
```
