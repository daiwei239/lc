```python
def getHappyString(self, n: int, k: int) -> str:
    res = []
    def dfs(n:int,curr:str):
        if n == 0:
            res.append(curr)
            return
        for ss in ["a","b","c"]:
            if not curr:
                curr=curr+ss
                dfs(n-1,curr)
                curr=curr[:-1]

            else:
                if ss!=curr[-1]:
                    curr=curr+ss
                    dfs(n-1,curr)
                    curr=curr[:-1]
    dfs(n,"")
    if len(res)>=k:
        return res[k-1]
    return ""
```

```python
def getHappyString(self, n: int, k: int) -> str:
    if k>3*2**(n-1):
        return ""
    res = ""
    prev = ""
    for i in range(n):
        block = 2**(n-i-1)
        for ss in ["a","b","c"]:
            if prev == ss:
                continue
            if k>block:
                k-=block
            else:
                res+=ss
                prev = ss
                break
        
    return res
```

```python
def findMaximizedCapital(self, k: int, w: int, profits: List[int], capital: List[int]) -> int:
    if w>=max(capital):
        return w+sum(nlargest(k,profits))
    
    n = len(profits)
    curr = 0
    arr = [(capital[i],profits[i]) for i in range(n)]
    arr.sort(key= lambda x:x[0])

    pq = []

    for _ in range(k):
        while curr<n and arr[curr][0]<=w:
            heapq.heappush(pq,-arr[curr][1])
            curr+=1
        if pq:
            w +=(-heapq.heappop(pq))
        else:
            break
    return w
        
```

# 题目**212**✨✨✨

```python
class Trie:
    def __init__(self):
        self.children = defaultdict(Trie)
        self.word = ""
        self.isWord = False
    def insert(self,word:str):
        curr = self
        for ss in word:
            curr = curr.children[ss]
        curr.word = word
        curr.isWord = True
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        trie = Trie()
        m = len(board)
        n = len(board[0])
        for word in words:
            trie.insert(word)
        ans = set()
        def dfs(now:Trie,i:int,j:int):
            nonlocal ans
            ch = board[i][j]

            if ch not in now.children:
                return 

            now = now.children[ch]
            
            if now.isWord == True:
                ans.add(now.word)
            board[i][j] = "#"

            for ii,jj in [(-1,0),(1,0),(0,1),(0,-1)]:
                x,y = i+ii,j+jj
                if 0<=x<m and 0<=y<n:
                    dfs(now,x,y)

            board[i][j] = ch
        for i in range(m):
            for j in range(n):
                dfs(trie,i,j)
        return list(ans)


            
```
