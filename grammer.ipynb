## str

```python
s = "hello"

s[0]    # 'h'
s[-1]   # 'o'
```

```python
for i, c in enumerate(s):
    print(i, c)
```

```python
s = "abcdef"

s[1:4]   # "bcd"
s[:3]    # "abc"
s[3:]    # "def"
```
