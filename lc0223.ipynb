# 题目**1461**✨✨

```python
def hasAllCodes(self, s: str, k: int) -> bool:
    if len(s)<(1<<k)+k-1:
        return False
    
    res = set( s[i:i+k] for i in range(len(s)-k+1))
    return len(res) == (1<<k)
```

# 题目**77**✨✨

```python
from typing import List
```

```python
def combine(self, n: int, k: int) -> List[List[int]]:
    res = []
    path = []
    def dfs(t:int):
        need = k -len(path)
        if need == 0:
            res.append(path.copy())
            return

        for i in range(t,need-1,-1):
            path.append(i)
            dfs(i-1)
            path.pop()
    return res
```

# 题目**46**✨✨

```python
def permute(self, nums: List[int]) -> List[List[int]]:
    res = []
    curr = []

    def dfs(k:int):
        if k == 0:
            res.append(curr.copy())
            return
    
        for i in nums:
            if i not in curr:
                curr.append(i)
                dfs(k-1)
                curr.pop()
    dfs(len(nums))
    return res

```

# 题目**39**✨✨

```python
def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
    res = []
    curr = []
    candidates.sort()
    def dfs(curTar:int,start:int)->None:
        if curTar == 0:
            res.append(curr.copy())
            return
         
        for i in range(start,len(candidates)):
            ii = candidates[i]
            if ii <= curTar:
                curr.append(ii)
                dfs(curTar-ii,i)
                curr.pop()
            else:
                break

    dfs(target,0)
    return res
```
