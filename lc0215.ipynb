# 题目**67**✨

```python
def addBinary(self, a: str, b: str) -> str:
    a = a[::-1]
    b = b[::-1]
    i,j =0,0
    last = 0
    res=[]
    while i<len(a) and j <len(b):
        temp = (int(a[i])+int(b[j])+ last)%2
        last = (int(a[i])+int(b[j])+ last)//2
        res.append(temp)
        i+=1
        j+=1
    while i<len(a):
        temp = (int(a[i])+ last)%2
        last = (int(a[i])+ last)//2
        res.append(temp)
        i+=1
    while j<len(b):
        temp = (int(b[j])+ last)%2
        last = (int(b[j])+ last)//2
        res.append(temp)
        j+=1
    if last:
        res.append(last)
    res.reverse()
    return ''.join(map(str,res))
```

# 题目**130**✨✨

```python
from typing import List
```

```python
def solve(self, board: List[List[str]]) -> None:
    m = len(board)
    n = len(board[0])
    # 处理边界上的0点
    def dfs(x:int,y:int):
        if not 0<=x<m or not 0<=y<n or board[x][y]=="X":
            return 
        board[x][y]="T"
        dfs(x+1,y)
        dfs(x-1,y)
        dfs(x,y-1)
        dfs(x,y+1)
    for i in range(0,m):
        dfs(i,0)
        dfs(i,n-1)
    for j in range(0,n):
        dfs(0,j)
        dfs(m-1,j)
    for i in range(m):
        for j in range(n):
            if board[i][j]=="T":
                board[i][j] = "O"
            elif board[i][j] =="O":
                board[i][j] = "X"
```
