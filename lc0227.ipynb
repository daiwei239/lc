# 题目**3666**✨✨✨

```python
from sortedcontainers import SortedList
```

```python
def minOperations(self, s: str, k: int) -> int:
    n = len(s)
    not_vis = [SortedList(range(0,n+1,2)),SortedList(range(1,n+1,2))]
    not_vis[0].append(n+1)
    not_vis[1].append(n+1)
    start = s.count("0")
    q = [start]
    not_vis[start%2].discard(start)
    ans = 0
    while q:
        temp = q
        q = []
        for z in temp:
            if z == 0:
                return ans
            min_val = z+k-2*min(k,z)
            max_val = z+k-2*max(0,z-n+k)
            sl = not_vis[min_val%2]
            idx = sl.bisect_left(min_val)

            while sl[idx]<=max_val:
                j = sl.pop(idx)
                q.append(j)
        ans+=1
    return -1
```

# 题目**34**✨✨

```python
from typing import List
```

```python
def searchRange(self, nums: List[int], target: int) -> List[int]:
    def find_low_bound(target:int)->int:
        left,right = 0,len(nums)-1
        while left<=right:
            mid = left+(right-left)//2
            
            if nums[mid]>=target:
                right = mid - 1
            else:
                left = mid + 1
        return left
    start = find_low_bound(target)
    if start == len(nums) or nums[start]!= target:
        return [-1,-1]
    end = find_low_bound(target+1)-1
    return [start,end]
```

# 题目**153**✨✨

>判断逻辑有点复杂

```python
def findMin(self, nums: List[int]) -> int:
    left,right = 0,len(nums)-1
    while left<=right:
        if nums[left]<nums[right]:
            return nums[left]
        else:
            mid = left+(right-left)//2
            if nums[mid]<nums[left] and nums[mid]<nums[right]:
                right = mid
            else:
                left = mid+1
    return nums[left-1]
```

## 方法二，简单的逻辑判断方式

```python
def findMin(self, nums: List[int]) -> int:
    left,right = 0,len(nums)-1
    while left<right:
        mid = left+(right-left)//2
        if nums[mid]<nums[right]:
            right = mid
        else:
            left = mid+1
    return nums[left]
```

# 题目**67**✨

```python
def addBinary(self, a: str, b: str) -> str:
    a = a[::-1]
    b = b[::-1]
    last = 0
    res = []
    i,j = 0,0
    while i <len(a) and j <len(b):
        temp = (int(a[i])+int(a[j])+last)%2
        last = (int(a[i])+int(a[j])+last)//2
        res.append(temp)
        i+=1
        j+=1
    
    while i<len(a):
        temp = (int(a[i])+last)%2
        last = (int(a[i])+last)//2
        res.append(temp)
        i+=1
    while j <len(b):
        temp = (int(b[j])+last)%2
        last = (int(b[j]+last))//2
        res.append(temp)
        j+=1
    if last:
        res.append(last)

    res.reverse()
    return ''.join(map(str,res))
```

# 题目**190**✨

```python
def reverseBits(self, n: int) -> int:
    temp = bin(n)[2:].zfill(32)
    temp = temp[::-1]
    return int(temp,2)
```
