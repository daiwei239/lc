# 题目**1653**✨✨

>本题的含义是，a存在的位置要比b存在的位置下标要小即可

```python
def minimumDeletions(self, s: str) -> int:
    n = len(s)

    if n<2:
        return 0
    dp = [0]*n
    countb = 0 #记录当前的下标下，前面有多少个b
    for i,j in enumerate(s):
        if j =="b":
            countb+=1
            dp[i]=dp[i-1] if i>1 else 0
        else:
            dp[i] = min(dp[i-1]+1,countb)
    return dp[n-1]
```

# 题目**20**✨

```python
from collections import deque
```

```python
def isValid(self, s: str) -> bool:
    stack = deque()
    for i in s:
        if i in ["(","[","{"]:
            stack.append(i)
            continue
        if len(stack)==0:
            return False
        if i ==")" and stack[-1]=="(":
            stack.pop()
        elif i =="}" and stack[-1]=="{":
            stack.pop()
        elif i == "]" and stack[-1]=="[":
            stack.pop()
        else:
            stack.append(i)
    return len(stack)==0
```

# 题目**71**✨✨

>这道题目用stack去进行表示    
>同时需要注意的是一个栈存文件名  
>然后开始遍历整体的  

```python
def simplifyPath(self, path: str) -> str:
    path = path.split("/")
    stack = list()
    for i in path:
        if i =="..":
            if stack:
                stack.pop()
        elif i and i !=".":
            stack.append(i)
    return "/" + "/".join(stack)
```

# 题目**155**✨✨

```python
import math
```

```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.minstack = [math.inf]

    def push(self, val: int) -> None:
        self.stack.append(val)
        self.minstack.append(min(self.minstack[-1],val))
        
        
    def pop(self) -> None:
        self.stack.pop()
        self.minstack.pop()
        
        

    def top(self) -> int:
        return self.stack[-1]
        
        
    def getMin(self) -> int:
        return self.minstack[-1]
        
```

# 题目**150**✨✨

```python
from typing import List
```

```python
def evalRPN(self, tokens: List[str]) -> int:
    stack = list()

    for c in tokens:
        if c not in "*/+-":
            stack.append(int(c))
        else:
            if stack:
                right = stack.pop()
                left  = stack.pop()
                if c =="*":
                    stack.append(left*right)
                elif c =="/":
                    stack.append(int(left/right))
                elif c =="+":
                    stack.append(left+right)
                elif c =="-":
                    stack.append(left-right)
    return stack[-1]

```

# 题目**141** ✨

```python
from typing import Optional
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
```

```python
def hasCycle(self, head: Optional[ListNode]) -> bool:
    slow ,fast = head,head
    while slow and fast and fast.next:
        if slow ==fast:
            return True
        slow =slow.next
        fast = fast.next.next
    return False
```

## 方法二
>使用hashtable的方式，进行单向存储

```python
def hasCycle(self, head: Optional[ListNode]) -> bool:
    hash = dict()
    while head:
        hash[head]= head
        if head.next in hash:
            return True
        head = head.next
    return False
```

# 题目**21**✨

```python
def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
    res = ListNode(0)
    curr = res
    while list1 and list2:
        if list1.val<list2.val:
            curr.next = list1
            list1 = list1.next
        else:
            curr.next = list2
            list2 = list2.next
        curr = curr.next
    if list1:
        curr.next = list1
    if list2:
        curr.next = list2
    return res.next
```
