# 题目**3546**✨✨

```python
from typing import List
```

```python
def canPartitionGrid(self, grid: List[List[int]]) -> bool:
    m = len(grid)
    n = len(grid[0])
    rows = [0]*m
    cols = [0]*n
    for i in range(m):
        rows[i] = sum(grid[i])
    for j in range(n):
        cols[j] = sum(grid[i][j] for i in range(m))
    total = sum(rows)
    curr = 0
    for i in range(m):
        curr+=rows[i]
        if curr == total//2:
            return True
        if curr>total//2:
            break
    curr = 0
    for j in range(n):
        curr+=cols[j]
        if curr == total//2:
            return True
        if curr>total//2:
            break
    return False
```

# 题目**207**✨✨

```python
from collections import deque,defaultdict
```

```python
def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
    orFinish = [0]*numCourses
    hash = defaultdict(list)
    for i,j in prerequisites:
        orFinish[i] +=1
        hash[j].append(i)


    queue = deque()
    for i,val in enumerate(orFinish):
        if i==0:
            queue.append(i)

    while queue:
        x = queue.popleft()
        for i in hash[x]:
            orFinish[i]-=1
            if orFinish[i] ==0:
                queue.append(i)
            
    if sum(orFinish)==0:
        return True
    
    return False
```

# 题目**208**✨✨

```python
class Trie:

    def __init__(self):
        self.children = [None]*26
        self.isEnd = False
        

    def insert(self, word: str) -> None:
        node = self
        for w in word:
            idx = ord(w)-ord('a')
            if node.children[idx] is None:
                node.children[ord('w')-ord('a')] = Trie()
            node = node.children[idx]
        node.isEnd = True
        

    def search(self, word: str) -> bool:
        node = self
        for w in word:
            if node.children[ord(w)-ord('a')]:
                node = node.children[ord(w)-ord('a')]
            else:
                return False
        return node.isEnd
        

    def startsWith(self, prefix: str) -> bool:
        node = self
        for w in prefix:
            if node.children[ord(w)-ord('a')]:
                node = node.children[ord(w)-ord('a')]
            else:
                return False
        return True
```

# 题目**39**✨✨

```python
def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
    res = []
    curr = []
    def dfs(target:int,start:int)->None:
        if target==0:
            res.append(curr.copy())
            return
        for i in range(start,len(candidates),1):
            if target<candidates[i]:
                continue
            curr.append(candidates[i])
            dfs(target-candidates[i],i)
            curr.pop()
    dfs(target,0)
    return res
        
```

# 题目**22**✨✨

```python
def generateParenthesis(self, n: int) -> List[str]:
    res = []
    curr = []
    def dfs(left:int,right:int)->None:
        nonlocal res
        nonlocal curr
        if left == 0 and right == 0:
            res.append("".join(curr))
            return
        if left>right:
            return
        if left>0:
            curr.append('(')
            dfs(left-1,right)
            curr.pop()
        if right>0:
            curr.append(')')
            dfs(left,right-1)
            curr.pop()
    dfs(n,n)
    return res
    

```

# 题目**79**✨✨

```python
def exist(self, board: List[List[str]], word: str) -> bool:
    length = len(word)
    m = len(board)
    n = len(board[0])
    orVisited = [[False]* n for _ in range(m)]
    def dfs(curr:int,i:int,j:int)->bool:
        nonlocal orVisited
        if word[curr] != board[i][j]:
            return False
        if word[curr]==board[i][j]:
            if curr == length-1:
                return True
            orVisited[i][j] = True
            for ii,jj in [(-1,0),(1,0),(0,1),(0,-1)]:
                x = i+ii
                y = j+jj
                if 0<=x<m and 0<=y<n and not orVisited[x][y] and dfs(curr+1,x,y):
                    return True
            orVisited[i][j] = False
        return False
    for i in range(m):
        for j in range(n):
            if dfs(0,i,j):
                return True

    return False
```

# 题目**131**✨✨

```python
def partition(self, s: str) -> List[List[str]]:
    n = len(s)
    f = [[True]*n for _ in range(n)]
    for i in range(n-1,-1,-1):
        for j in range(i,n):
            f[i][j] = f[i+1][j-1] and f[i][j]

    res = []
    onetime = []
    def dfs(curr:int):
        nonlocal res
        nonlocal onetime
        if curr == n:
            res.append(onetime[:])
            return
        for j in range(curr,n):
            if f[i][j]:
                onetime.append(s[i:j+1])
                dfs(j+1)
                onetime.pop()
    dfs(0)
    return res
```
