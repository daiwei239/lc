# 题目**110**✨

>直接进行判定，
>用深度搜索，如果当前左右子树的高度差值大于1了  
>就判定其为不平衡的子树，否则就认为是平衡的子树

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
from typing import Optional
```

```python
def isBalanced(self, root: Optional[TreeNode]) -> bool:
    inf = float("inf")
    def dfs(root:Optional[TreeNode])->int:
        if root==None:
            return 0
        leftl = dfs(root.left)
        rightl = dfs(root.right)
        if abs(rightl-leftl)>1:
            return inf
        return max(leftl,rightl)+1
    l = dfs(root)
    return False if l ==inf else True
 
```

# 题目**138**✨✨

```python

# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
```

>本题使用新节点和原本节点交替构造的思想
>先构建出原节点的交替新节点，然后再去 
>以此构建random节点，最后将原节点其删除掉即可

```python
def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
    if head ==None:
        return None
    newHead = Node(-1)
    newHead.next = head
    while head:
        temp = head.next
        no = Node(head.val)
        head.next = no
        no.next = temp
        head = temp
    # 现在构造出来的节点顺序是
    # newHead ->head->新节点-》旧节点-》新节点-》旧节点
    head = newHead.next

    # 构建random值
    while head:
        head.next.random = head.random.next
        head = head.next.next
    # 修剪结果，返回值
    temp = newHead.next.next
    newHead.next = temp
    while temp :
        temp.next = temp.next.next
    return newHead.next
```

# 题目**92**✨✨

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
    if head is None:
        return None
    dummy = ListNode(-1,next=head)
    prev = dummy
    i=0
    while i< (left-1):
        prev = prev.next
        i+=1
   
    
    curr =prev.next
    while i<right-1:
        nxt = curr.next
        curr.next = nxt.next
        nxt.next = prev.next
        prev.next = nxt
        i+=1
        
    return dummy.next
        
```

# 题目**19**✨✨

```python
def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
    dummy = ListNode(-1,head)
    counter = 0
    while head :
        counter+=1
        head = head.next
    
    head = dummy.next
    for _ in range(counter-n-1):
        head = head.next
    head.next = head.next.next

    return dummy.next
```

# 题目**82**✨✨

```python
def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
    dummy = ListNode(-1,head)
    res = dummy
    fir = head
    
    while fir:
        sec = fir.next
        while sec and sec.val ==fir.val:
            sec = sec.next
        if fir.next == sec:
            res.next =fir
            res = res.next
        fir = sec
    
    res.next = None
    return dummy.next
```
