# 题目**108**✨

```python
from typing import List,Optional
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
    def build (left:int,right:int)->Optional[TreeNode]:
        if left > right:
            return None
        mid = (left+right)//2
        root = TreeNode(nums[mid])
        root.left = build(left,mid-1)
        root.right = build(mid+1,right)
        return root
    return build(0,len(nums)-1)
```

# 题目**1382**✨✨

```python
def balanceBST(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
    res = list()
    def inorderTraversal(root:Optional[TreeNode])->None:
        if root is None:
            return
        inorderTraversal(root.left)
        res.append(root.val)
        inorderTraversal(root.right)
    
    def build(left:int,right:int)->Optional[TreeNode]:
        if left>right:
            return None
        mid = (left+right)//2
        root = TreeNode(res[mid])
        root.left = build(left,mid-1)
        root.right = build(mid+1,right)
        return root
    inorderTraversal(root)
    return build(0,len(res)-1)
        
```

# 题目**61**✨✨

```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
    if head is None:
        return head
    counter = 0
    fir = head
    while fir:
        fir = fir.next
        counter+=1
    k %=counter
    k= counter-k
    fir = head
    for _ in range(k-1):
        fir = fir.next
    res = fir.next
    fir.next = None
    fir = res 
    
    while fir.next:
        fir = fir.next
    fir.next = head
    return res
```

# 题目**86**✨✨

```python
def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
    small_dummy,big_dummy = ListNode(-1),ListNode(-1)
    small = small_dummy
    big = big_dummy
    while head:
        if head.val>=x:
            big.next = head
            big = big.next
        else:
            small.next = head
            small = small.next
        head = head.next
    small.next = big_dummy.next
    big.next = None
    return small_dummy.next
```

# 题目**146**✨✨

```python
from collections import OrderedDict
```

```python
class LRUCache:

    def __init__(self, capacity: int):
        self.capicity = capacity
        self.table = OrderedDict()

    def get(self, key: int) -> int:
        if key in self.table:
            self.table.move_to_end(key)
        return self.table[key]
        
        

    def put(self, key: int, value: int) -> None:
        self.table[key]=value
        self.table.move_to_end(key)
        if len(self.table)>self.capicity:
            self.table.popitem(last=False)

        

        


# Your LRUCache object will be instantiated and called as such:
# obj = LRUCache(capacity)
# param_1 = obj.get(key)
# obj.put(key,value)
```
