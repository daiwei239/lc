```python
from typing import List
```

```python
def getBiggestThree(self, grid: List[List[int]]) -> List[int]:
    m,n = len(grid),len(grid[0])
    diag_sum = [[0]*(n+1) for _ in range(m+1)]
    anti_sum = [[0]*(n+1) for _ in range(m+1)]
    for i,row in enumerate(grid):
        for j,v in enumerate(row):
            diag_sum[i+1][j+1] = diag_sum[i][j]+v
            anti_sum[i+1][j] = anti_sum[i][j+1]+v
    
    def query_diagnoal(x:int,y:int,k:int):
        return diag_sum[x+k][y+k]-diag_sum[x][y]
    
    def query_anti(x:int,y:int,k:int):
        return anti_sum[x+k][y+1-k]-anti_sum[x][y+1]
    
    x=y=z=0

    def update(newV:int):
        nonlocal x,y,z
        if newV>x:
            x,y,z = newV,x,y
        elif x>newV>y:
            y,z = newV,y
        elif y>newV>z:
            z = new=

    for i,row in enumerate(grid):
        for j ,v in enumerate(row):
            update(v)

            mx = (i,m-1-i,j,n-1-j)

            for k in range(1,mx+1):
                a = query_diagnoal(i-k,j,k)
                b = query_diagnoal(i,j-k,k)
                c = query_anti(i-k+1,j-1,k-1)
                d = query_anti(i,j+k,k+1)
                update(a+b+c+d)
    ans = [x,y,z]

    while ans[-1] == 0:
        ans.pop()
    return ans
```
