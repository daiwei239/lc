```python
def longestBalanced(self, nums: List[int]) -> int:
    ans = 0
    for i in range(len(nums)):
        odd_dict = 0
        even_dict = 0
        hash = dict()
        for j in range(i,len(nums)):
            if nums[j] not in hash:
                hash[nums[j]]=1
                if nums[j]%2 ==0:
                    even_dict+=1
                else:
                    odd_dict+=1
            if even_dict==odd_dict:
                ans = max(ans,j-i+1)
    return ans
```

# 题目**104**✨

```python
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def maxDepth(self, root: Optional[TreeNode]) -> int:
    def dep(root:Optional[TreeNode])->int:
        if root is None:
            return 0
        left = dep(root.left)
        right = dep(root.right)
        return max(left,right)+1
    return dep(root)
```

# 题目**100**✨


```python
def isSameTree(self, p: Optional[TreeNode], q: Optional[TreeNode]) -> bool:
    def same(p: Optional[TreeNode], q: Optional[TreeNode])->bool:
        if p is None and q is None:
            return True
        if p.val != q.val:
            return False

        left = same(p.left,q.left)
        right = same(p.right,q.right)
        return left and right
    return same(p,q)
    
```

# 题目**226**✨

```python
def invertTree(self, root: Optional[TreeNode]) -> Optional[TreeNode]:
    def invert(root:Optional[TreeNode])->TreeNode:
        if root is None:
            return 
        left = invert(root.left)
        right = invert(root.right)
        root.left =right
        root.right = left
        return root
    return invert(root)
    
```

```python
def isSymmetric(self, root: Optional[TreeNode]) -> bool:
    def compare(left:Optional[TreeNode],right:Optional[TreeNode])->bool:
        if not left and not right:
            return True
        if left and right and left.val != right.val:
            return False
        l = False
        r = False
        if left and right:
            l = compare(left.right,right.left)
            r = compare(left.left,right.right)

        return l and r
        
    if not  root:
        return False
    return compare(root.left,root.right)
```

# 题目**105**✨✨

```python
def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
    
    in_map = {val: i for i, val in enumerate(inorder)}
    def pi(prest,preend,inst,inend)->Optional[TreeNode]:
        if prest>preend or inst>inend:
            return None
        
        root = TreeNode(preorder[prest])
        i = in_map[preorder[prest]]
        left_size = i-inst
        root.left = pi(prest+1,prest+left_size,inst,i-1)
        root.right = pi(i+1,preend,i+1,inend)
        return root
    return pi(0,len(preorder),0,len(inorder))
```
