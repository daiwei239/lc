```python
def isTrionic(self, nums: List[int]) -> bool:
    n = len(nums)
    right = 0
    left = 0
    while right+1<n and nums[right]<nums[right+1]:
            right+=1
    # 从现在开始不是递增的了
    if right==left:
        return False    
    left = right
    while right+1<n and nums[right]>nums[right+1]:
        right+=1
    # 从现在开始不是递减的了
    if right==left:
        return False
        
    left = right
    while right+1<n and nums[right]<nums[right+1]:
        right+=1
    if right==left:
        return False
    return right==n-1
```

# 题目**11**✨✨

```python
def maxArea(self, height: List[int]) -> int:
    left,right = 0,len(height)-1
    res = 0

    while left<right:
        ht = min(height[left],height[right])
        res = max(res,ht*(right-left))
        if height[left]<height[right]:
            left+=1
        else:
            right-=1
    return res
```

# 题目**15**✨✨

```python
def threeSum(self, nums: List[int]) -> List[List[int]]:
    n = len(nums)
    nums.sort()
    res = []
    for i in range(n):
        if i>0 and nums[i]==nums[i-1]:
            continue
        target = -nums[i]
        left = i+1
        right = n-1
        while left<right:
            temp = nums[left]+nums[right]
            if temp<target:
                left+=1
            elif temp>target:
                right-=1
            else:
                res.append([nums[i],nums[left],nums[right]])
                while left<right and nums[left]==nums[left+1]:
                    left+=1
                while left< right and nums[right-1] == nums[right]:
                    right-=1
                left+=1
                right-=1
                
    return res
```

# 题目**209**✨✨

```python
def minSubArrayLen(self, target: int, nums: List[int]) -> int:
    left = 0
    curr = 0
    ans = float("inf")
    for right in range(len(nums)):
        curr+=nums[right]
        while curr>=target:
            ans = min(ans,right-left+1)
            left+=1
    return ans if sum(nums)>=target else 0

    
        
```

# 题目**3**✨✨

```python
def lengthOfLongestSubstring(self, s: str) -> int:
    maxlen = 0
    left = 0
    hash = {}
    for right in range(len(s)):
        if hash.get(s[right],0)==0:
            hash[s[right]] = right+1
        else:
            maxlen = max(maxlen,right-left)
            left =max(left,hash[s[right]])
            hash[s[right]]=right+1
    return max(maxlen,len(s)-left)
```

# 题目**383**✨

```python
def canConstruct(self, ransomNote: str, magazine: str) -> bool:
    hash = dict()
    for i in magazine:
        hash[i] = hash.get(i,0)+1
    
    for i in ransomNote:
        hash[i] =hash.get(i,0)-1
        if hash[i]<0:
            return False
    return True
```
