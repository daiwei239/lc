# 题目**3474**✨✨✨

```python
def generateString(self, str1: str, str2: str) -> str:
    m = len(str1)
    n = len(str2)
    ans = ['?']*(m+n-1)
    for i,val in enumerate(str1):
        if val == 'F':
            continue
        for j,c in enumerate(str2):
            vv = ans[i+j]
            if vv != '?' and vv != c:
                return ""
            ans[i+j] = c
    
    old_ans = ans
    ans = ['a' if c == '?' else c for c in ans]
    for i,val in enumerate(str1):
        if val == 'T':
            continue
        if ''.join(ans[i:i+n])!= str2:
            continue
        for j in range(i+n-1,i-1,-1):
            if old_ans[j]=='?':
                ans[j] = 'b'
                break
        else:
            return ""
    return "".join(ans)
        
```

# 题目**416**✨✨

```python
from typing import List
from collections import defaultdict
```

```python
def canPartition(self, nums: List[int]) -> bool:
    total = sum(nums)
    if total%2 !=0:
        return False
    total//=2
    dp = [False]*(total+1)
    dp[0] = True

    for i,vi in enumerate(nums):
        for j in range(total,vi-1,-1):
            dp[j] = dp[j] or dp[j-vi]
        if dp[total]:
            return True

    return dp[total]
```

# 题目**62**✨✨

```python
def uniquePaths(self, m: int, n: int) -> int:
    dp = [[1]*n for _ in range(m)]
    for i in range(1,m):
        for j in range(1,n):
            dp[i][j] = dp[i-1][j] +dp[i][j-1]
    
    return dp[m-1][n-1]
```

# 题目**64**✨✨

```python
def minPathSum(self, grid: List[List[int]]) -> int:
    m = len(grid)
    n = len(grid[0])
    dp = [[0]*n for _ in range(m)]
    
    
    for i in range(0,m):
        for j in range(0,n):
            if i==0 and j == 0:
                dp[0][0] = grid[0][0]
                continue
            if i == 0:
                dp[i][j] = dp[i][j-1]+grid[i][j]
                continue
            if j == 0:
                dp[i][j] = dp[i-1][j] +grid[i][j]
                continue
            
            dp[i][j] = min(dp[i-1][j],dp[i][j-1])+grid[i][j]
    return dp[m-1][n-1]
```

# 题目**5**✨✨

```python
def longestPalindrome(self, s: str) -> str:
    n = len(s)
    if n < 2:
        return s
    dp = [[False]* n for _ in range(n)]
    for i in range(n):
        dp[i][i]= True
    
    for i in range(n-1,-1,-1):
        for j in range(i,n,1):
            if i+1<=j-1:
                dp[i][j] = dp[i+1][j-1] and s[i]==s[j]
            else:
                dp[i][j] = s[i]==s[j]

    st = 0
    en = 0
    maxlen = 0
    for i in range(n):
        for j in range(n):
            if dp[i][j] and (j-i+1)>maxlen:
                maxlen = j-i+1
                st = i
                en = j
    return s[st:en+1]
```

# 题目**1143**✨✨

```python
def longestCommonSubsequence(self, text1: str, text2: str) -> int:
    m = len(text1)
    n = len(text2)
    dp =[[0]*(m+1) for _ in range(n+1)]
    for i in range(1,m+1):
        for j in range(1,n+1):
            if text1[i-1] ==text2[j-1]:
                dp[i][j] = dp[i-1][j-1]+1
            else:
                dp[i][j] = max(dp[i-1][j],dp[i][j-1])
    return dp[m][n]
```
