```python
def minimumPairRemoval(self, nums: List[int]) -> int:
    # 直接进行模拟遍历操作
    # 一直循环到结束为止
    counter = 0
    while len(nums)>1:
        ascending = True
        minnum = float("inf")
        changeIndex = -1
        for i in range(len(nums)-1):

            if nums[i]>nums[i+1]:
                ascending = False

            if minnum>nums[i]+nums[i+1]:
                minnum = nums[i]+nums[i+1]
                changeIndex=i
        if ascending is True:
            break
        # 当前的数组还是需要进行改变
        nums[changeIndex]=minnum
        nums.pop(changeIndex+1)
    
    return counter
```

```python
def firstMissingPositive(self, nums: List[int]) -> int:
    # 对于这一题，先将小于等于0的值设置为len(nums)
    # 这个时候原本的值全部变为了正数
    # 然后对于数组开始进行遍历,将其自身看成是hash表
    # 重要的是对自身进行标记，用负号进行标记
    # 规则是在遍历第i个元素时,将nums[i]位置上的值改成原本值的-数
    # 最后研判一下哪个位置上没有出现负值
    nums.append(0)
    n=len(nums)
    for i in range(len(nums)):
        if nums[i]<=0:
            nums[i]=n

    for ij in nums:
        if abs(ij)<n:
            nums[abs(ij)]=-abs(nums[abs(ij)])

    for i in range(1,n):
        if nums[i] > 0:
                
            return i
            
    return n
```

```python
# 采用类似归并排序的方式
from typing import Optional
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
def mergeTwo(self,list1:Optional[ListNode],list2:Optional[ListNode])->Optional[ListNode]:
        merged=ListNode(-1)
        dummy = merged
        
        while list1 and list2:
            if list1.val < list2.val:
                merged.next=list1
                list1 = list1.next
            else:
                merged.next=list2
                list2 = list2.next
            merged = merged.next
        if list1:
            merged.next = list1
        else:
            merged.next = list2

        return dummy.next


# 对多个数组进行归并排序
def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
    if not lists:
        return []
    while len(lists)>1:
        merged=[]
        # for i in range(len(lists)):
        for i in range(0,len(lists),2):
            if i+1<len(lists):
                merged.append(mergeTwo(lists[i],lists[i+1]))
            else:
                merged.append(lists[i])
        lists = merged
    # return lists
    return lists[0]
```

```python
def sortArray(self, nums: List[int]) -> List[int]:

    def mergeTwo(nums1:List[int],nums2:List[int])->List[int]:
        len1 = len(nums1)
        len2 = len(nums2)
        i,j = 0,0
        merged = []
        while i<len1 and j<len2:
            if nums1[i]<nums2[j]:
                merged.append(nums1[i])
                i+=1
            else:
                merged.append(nums2[j])
                j+=1
        merged.append(nums1[i:] if nums1[i] else nums2[j:])
        return merged
    
    # left,right = 0,len(nums)-1 这种写法 right会漏掉一个边界值
    left,right = 0,len(nums)

    mid = (left+right)//2
    leftNums =sortArray(nums[0:mid])
    rightNums =sortArray(nums[mid+1:])
    return mergeTwo(leftNums,rightNums) 
```
