# 题目**3013**✨✨✨

```python
from typing import List
from sortedcontainers import SortedList
```

```python
def minimumCost(self, nums: List[int], k: int, dist: int) -> int:
    k-=1
    Left = SortedList(nums[1:dist+2])
    Right = SortedList()
    sumL = sum(nums[:dist+2])
    
    def L2R()->None:
        x = Left.pop()
        nonlocal sumL
        sumL -= x
        Right.add(x)

    def R2L()->None:
        x = Right.pop(0)
        nonlocal sumL
        sumL+=x
        Left.add(x)

    while len(Left)>k:
       L2R()
    ans = sumL#第一个初始轮的ans

    for i in range(dist+2,len(nums)):
        out_value = nums[i-dist-1]

        # 如果在left里面，就把left内容弹出去
        # sumL-=out
        # 如果在right，就操作right

        if out_value in Left:
            Left.remove(out_value)
            sumL-=out_value
        else:
            Right.remove(out_value)
        
        in_value = nums[i]
        # 如果 in_value <Left[-1]
        # 就加入Left
        # 否则就直接加入right
        if in_value <Left[-1]:
            Left.add(in_value)
            sumL+=in_value
        else:
            Right.add(in_value)
        
        # 对于Left和Right的大小进行维护
        # 为什么要在这操作
        # 如果left弹出去一个后，in_value>L[-1]
        # 那么left就亏损了。而非一直是len(left)==k(这个k被调整过了)
        if len(Left)==k-1:
            R2L()
        if len(Left)==k+1:
            L2R()
        
        ans = min(ans,sumL)

    return ans
```

# 题目**6**✨✨

>本题直接使用数值变换对应不同的char位置，实现  
>填充，然后再将其遍历即可

```python
def convert(self, s: str, numRows: int) -> str:
        if numRows ==1:
            return s
        n = len(s)
        columns = n
        table = [[" "]*columns for _ in range(numRows)]

        i,j =0,0
        cycle = 2*numRows-2
        for t in range(n):
            table[i][j] = s[t]
            # i = i+1 if t % 4 < 2 else i-1 
            # j = j+1 if t % 5 >2 else j
            pos = t%cycle
            if pos <numRows-1:
                i+=1
            else:
                i-=1
                j+=1
        
        res = ""
        for i in range (numRows):
            for j in range(columns):
                if table[i][j] != " ":
                    res+=table[i][j]
        
        return res
```

## 方法二（官方解答）
>使用第一种方式的做法，时间复杂度过高  
>法二更加有思考性也更加巧妙


```python
def convert(self, s: str, numRows: int) -> str:
    res = ["" for _ in range(numRows)]
    i,flag =0,-1
    for c in s:
        res[i]+=c
        if i ==0 or i ==numRows-1:
            flag = -flag
        i+=flag
    return "".join(res)
```

# 题目**28**✨

```python
def strStr(self, haystack: str, needle: str) -> int:
    m = len(haystack)
    n = len(needle)
    i,j = 0,0
    if m<n:
        return -1
    while i<m:
        while j <n and i<m:
            
            if haystack[i] == needle[j]:
                i+=1
                j+=1
                if j==n:
                    return i-j
            else:
                i=i-j+1
                j = 0
        if j==0:
            i+=1
    return -1
            
```

# 题目**125**✨

```python
def isPalindrome(self, s: str) -> bool:
    if len(s)<2:
        return True

    left,right = 0,len(s)-1
    while left<right:
        while not s[left].isalnum():
            left+=1
        while not s[right].isalnum():
            right-=1
        if s[left].lower()==s[right].lower():
            left+=1
            right-=1
        else:
            return False
    return True
```

# 题目**392**✨

```python
def isSubsequence(self, s: str, t: str) -> bool:
    m,n = len(s),len(t)
    if m ==0 :
        return True
    if m>n:
        return False
    i,j =0,0
    while j<n:          
        if s[i]==t[j]:
            i+=1
        if i>=m:
            return True
        j+=1
    return False
```

# 题目**167**✨✨

```python
def twoSum(self, numbers: List[int], target: int) -> List[int]:
    left ,right = 0,len(numbers)-1
    while left<right:
        if numbers[left]+numbers[right]<target:
            left+=1
        elif numbers[left]+numbers[right]>target:
            right-=1
        elif numbers[left]+numbers[right]==target:
            return [left,right]
```
