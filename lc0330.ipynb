# 题目**2840**✨✨

>通过验证，用单数和双数的hash，用来存储这些内容，如果两者最后都分配为相同的内容，就是可以作为结果的，否则就是错误的

```python
from collections import defaultdict
```

```python
def checkStrings(self, s1: str, s2: str) -> bool:
    evenhash = defaultdict(int)
    oddhash = defaultdict(int)
    for i in range(len(s1)):
        if i %2 !=0:
            oddhash[s1[i]] = oddhash.get(s1[i])+1
        else:
            evenhash[s1[i]] = evenhash.get(s1[i])+1
    
    for i in range(len(s2)):
        if i % 2 != 0:
            if s2[i] not in oddhash:
                return False
            oddhash[s2[i]] = oddhash.get(s2[i])-1
        else:
            if s2[i] not in evenhash:
                return False
            evenhash[s2[i]] = evenhash.get(s2[i])-1
    
    for k,v in oddhash.items():
        if v !=0:
            return False
    for k,v in evenhash.items():
        if v != 0:
            return False
    return True
```

# 题目**279**✨✨

```python
from math import sqrt


def numSquares(self, n: int) -> int:
    dp = [float('inf')]*(n+1)
    dp[0] = 0
    for i in range(1,n+1):
        for ii in range(sqrt(i),0,-1):
            dp[i] = min(dp[i],dp[i-ii*ii]+1)
    return dp[n]
```

# 题目**139**✨✨

```python
from typing import List
```

```python
def wordBreak(self, s: str, wordDict: List[str]) -> bool:
    n = len(s)
    dp = [False]*(n+1)
    dp[0]= True
    wordDict = set(wordDict)
    for right in range(n):
        for left in range(right+1):
            if s[left:right+1] in wordDict and dp[left]:
                dp[right+1] = True
                break
    return dp[n]
```

# 题目**300**✨✨

```python
def lengthOfLIS(self, nums: List[int]) -> int:
    n = len(nums)
    dp = [1]*n
    
    for right in range(1,n):
        for i in range(right):
            if nums[i]<nums[right]:
                dp[right] = max(dp[i]+1,dp[right])
    
    return dp[n-1]
```

# 题目**152**✨✨

```python
def maxProduct(self, nums: List[int]) -> int:
    n = len(nums)
    dp = [[-float('inf'),float('inf')] for _ in range(n)]
    
    dp[0][0] = nums[0] 
    dp[0][1] = nums[0]
    for i in range(1,n):
        
            dp[i][0] = max(nums[i],dp[i-1][1]*nums[i],dp[i-1][0]*nums[i])
            dp[i][1] = min(nums[i],dp[i-1][1]*nums[i],dp[i-1][0]*nums[i])
    return max(dp[i][0] for i in range(n))
```
