# 题目**2751**✨✨✨

```python
from typing import List
```

```python
def survivedRobotsHealths(self, positions: List[int], healths: List[int], directions: str) -> List[int]:
    n = len(positions)
    seq = [i for i in range(n)]
    seq.sort(key=lambda i:positions[i])

    stack = list()

    for curr in seq:
        currIdx,currHealth,currDirection = curr,healths[curr],directions[curr]
        orPut = True
        while stack:
            stackIdx,stackHealth,stackDirection = stack[-1]
            if stackDirection =='R' and currDirection=='L':
                stack.pop()
                if stackHealth>currHealth:
                    stackHealth-=1
                    stack.append((stackIdx,stackHealth,stackDirection))
                    orPut = False
                    break
                elif currHealth == stackHealth:
                    orPut = False
                    break
                else:
                    currHealth-=1
            else:
                break
        if orPut:
            stack.append((currIdx,currHealth,stackDirection))

    stack.sort(key= lambda o:o[0])
    return [a[1] for a in stack]
```

# 题目**763**✨✨

```python
def partitionLabels(self, s: str) -> List[int]:

    if not s:
        return []
    hash = {}
    
    for i,val in enumerate(s):
        if val not in hash:
            hash[val]=[i,i]
        else:
            hash[val][1]=i
    hash = list(hash.values)
    hash.sort(key= lambda x:x[0])
    res = []
    left = hash[0][0]
    right = hash[0][1]
    for i in range (1,len(hash)):
        v = hash[i]
        if right<v[0]:
            res.append(right-left+1)
            left=v[0]
            right = v[1]
        else:
            right = max(right,v[1])

    res.append(right-left+1)
    return res
```

# 题目**33**✨✨

```python
def search(self, nums: List[int], target: int) -> int:
    left,right = 0,len(nums)-1
    while left<=right:
        mid = left+(right-left)//2
        if nums[mid] == target:   # 修改处：先判断是否找到
            return mid
        
        if nums[left]<=nums[mid]:
            if nums[left]<=target<nums[mid]:
                right = mid-1
            else:
                left = mid+1
        else:
            if nums[mid]<target<=nums[right]:
                left = mid+1
            else:
                right = mid-1
                
    return -1
```

# 题目**153**✨✨

```python
def findMin(self, nums: List[int]) -> int:
    left,right = 0,len(nums)-1
    while left< right:
        mid = left+(right-left)//2
        if nums[mid]<nums[right]:
            right=mid
        else:
            left = mid+1
    
    return nums[right]
```

# 题目**347**✨✨

```python
import heapq
```

```python
def topKFrequent(self, nums: List[int], k: int) -> List[int]:
    hash = {}
    for i,val in enumerate(nums):
        if val not in hash:
            hash[val] = 1
        else:
            hash[val] = hash.get(val)+1
    one = list()
    heapq.heapify(one)
    for key,value in hash.items():
        if len(one)<k:
            heapq.heappush(one,(value,key))
        else:
            topofheap = one[0][0]
            if value>topofheap:
                heapq.heappop(one)
                heapq.heappush(one,(value,key))
    return [i[1] for i in one]
```

# 题目**215**✨✨

```python
from random import randint


def findKthLargest(self, nums: List[int], k: int) -> int:
    n = len(nums)
    target = n-k
    left,right = 0,n-1
    while left<=right:
        l = left
        r = right
        pivot = randint(l,r)
        temp = nums[pivot]
        while l<=r:

            while l<=r and nums[l]<=temp:
                l+=1
            while l<=r and nums[r]>temp:
                r-=1
            if l<=r:
                nums[l],nums[r]= nums[r],nums[l]
                l+=1
                r-=1
        if l == target:
            return nums[target]
        if l<target:
            left = l+1
        else:
            right = r
    return nums[left]
            
```

# 题目**912**✨✨

```python

def sortArray(self, nums: List[int]) -> List[int]:
    def partition(left:int,right:int)->int:
        if left>right:
            return 
        temp = nums[randint(left,right)]
        l,r = left,right
        while l<=r:
            while l<=r and nums[l]<temp:
                l+=1
            while l<=r and nums[r]>temp:
                r-=1
            if l<=r:
                nums[l],nums[r] = nums[r],nums[l]
                l+=1
                r-=1
        return l
    

    def quickSort(left:int,right:int)->None:
        ordered = True
        for i in range(left,right):
            if nums[i]>nums[i+1]:
                ordered = False
        if ordered:
            return 
        i = partition(left,right)
        quickSort(left,i-1)
        quickSort(i+1,right)

    
    quickSort(0,len(nums)-1)
    return nums

    


```
