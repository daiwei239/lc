# 题目**3721**✨✨✨

```python
from typing import List
from collections import defaultdict
from math import isqrt
```

```python
fmax = lambda x,y :x if x>y else y
fmin = lambda x,y :x if x<y else y

class Solution:
    def longestBalanced(self,nums):

        # 初始化基础变量
        n = len(nums)

        # pre[i] 表示当前左端点为0时，去重后的奇偶平衡值
        pre = [0]*n

        p = 0

        mx = 0
        # pos[x] : 记录每个数字出现的位置列表
        # 用于快速找到“下一次出现位置”
        pos = defaultdict(list)
        for i,x in enumerate(nums):
            # 如果x第一次出现
            if x not in pos:
                # 奇数加1
                if x&1:
                    p+=1
                else:
                    p-=1
            pre[i] =p
            pos[x].append(i)
            
            if p ==0:
                mx = i+1

        # 进行根号n的分块结构

        size = isqrt(n)+1

        cb = (n+size-1)//size

        # cnt[b] :
        # 第 b 块中所有 pre[i] 的集合(set)
        #
        # 作用：
        #   快速判断这一块是否“可能存在0”
        #
        # 注意：
        #   cnt 存的是原始 pre 值
        #   不包含 tag 偏移
        #
        cnt = [set(pre[i*size:(i+1)*size]) for i in range(cb)]
        
        # tag[b] :
        # 第 b 块的懒更新偏移量
        #
        # 真实值 = pre[i] + tag[b]
        #
        tag = [0]*len(cnt)
        # -------------------------------
        # 四、区间更新函数 up()
        #
        # 功能：
        #   把 pre[l..r] 全部加上 x
        #
        # 思想：
        #   左碎块 + 中整块(tag) + 右碎块
        # -------------------------------
        def up(l,r,x):
            bl = l//size
            br = r//size

            if bl == br:
                for i in range(l,r+1):
                    pre[i]+=x
                cnt[bl]= set(pre[bl*size:(bl+1)*size])
            else:
                for i in range(l,(bl+1)*size):
                    pre[i]+=x
                cnt[bl] = set(pre[bl*size:(bl+1)*size])
                # 中间的完整块
                for i in range(bl+1,br):
                    tag[i]+=x
                
                for i in range(br*size,r+1):
                    pre[i]+=x
                cnt[br] = set(pre[br*size:(br+1)*size])
        # -------------------------------
        # 五、query函数
        #
        # 功能：
        #   在区间[l,r]中
        #   找最右侧满足
        #
        #       pre[i] + tag[block] == 0
        #
        #   的位置
        # -------------------------------

        def query(l,r):
            bl =l//size
            br = r//size

            if bl ==br:
                for i in range(r,l-1,-1):
                    if pre[i]+tag[bl]==0:
                        return i
                return -1
            else:
                # 如果再最右边那个散块
                for i in range(r,fmax(l,br*size)-1,-1):
                    if pre[i] +tag[br]==0:
                        return i
                # 如果是中间的整块
                for b in range(br-1,bl,-1):
                    if -tag[b] in cnt[b]:
                        x = tag[b]
                        
                        st = fmin((b+1)*size-1,n-1)
                        end = fmax(b*size,0)

                        for i in range(st,end-1,-1):
                            if pre[i]+x == 0:
                                return i
                
                for i in range(fmin((bl+1)*size,n)-1,l-1,-1):
                    if pre[i]+tag[bl]==0:
                        return i 
            return -1
        
        # -------------------------------
        # 六、主循环
        #
        # 枚举左端点 i
        # -------------------------------
        p = mx-1

        for i in range(n-1):
            x = nums[i]

            pos[x].pop(0)

            r = n if not pos[x] else pos[x][0]

            d = -1 if x&1 else 1

            up(i,r-1,d)

            if r<p:
                continue

            p = query(i+1,n-1)
            mx = fmax(mx,p-i)

            if mx>=(n-i):
                break
        return mx
```

# 题目**112**✨

```python
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
    def dfs(root:Optional[TreeNode],targetSum: int)->bool:
        if not root:
            return False
        if root.val ==targetSum and not root.left and not root.right:
            return True
        left = dfs(root.left,targetSum-root.val)
        right = dfs(root.right,targetSum-root.val)
        return left or right
    return dfs(root,targetSum)
```

# 题目**236**✨✨

```python
def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
    if root==None:
        return None
    if root ==p:
        return p
    if root ==q:
        return q
    left = lowestCommonAncestor(root.left,p,q)
    right = lowestCommonAncestor(root.right,p,q)
    if not left and not right:
        return root
    return left if left else right
```
