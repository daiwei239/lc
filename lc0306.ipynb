# 题目**1784**✨✍️

```python
def checkOnesSegment(self, s: str) -> bool:
    n = len(s)
    if n<2:
        return True
    last = -1
    for i in range(n):
        if s[i]=="1":
            if last ==i-1 or last == -1:
                last = i
            elif last != i-1:
                return False
    return True
```

# 题目**97**✨✨

```python
def isInterleave(self, s1: str, s2: str, s3: str) -> bool:
    if len(s1)+len(s2)!= len(s3):
        return False
    n1 = len(s1)
    n2 = len(s2)
    dp = [[False]*(n1+1) for _ in range(n2+1)]
    dp[0][0] = True
    for i in range(n1+1):
        for j in range(n2+1):
            if j> 0:
                dp[i][j] = dp[i][j] or (dp[i][j-1] and s2[j-1] == s3[i+j-1])
            if i > 0:
                dp[i][j] = dp[i][j] or (dp[i-1][j] and s1[i-1] == s3[i+j-1])
    return dp[n1][n2]

```

# 题目**72**✨✨

```python
def minDistance(self, word1: str, word2: str) -> int:
    m= len(word1)
    n = len(word2)
    dp = [[0]*(n+1) for _ in range(m+1)]
    for j in range(n+1):
        dp[0][j] = j
    for i in range(m+1):
        dp[i][0] = i
    for i in range(1,m+1):
        for j in range(1,n+1):
            if  word1[i-1] == word2[j-1]:
                dp[i][j] = min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1]-1)+1
            else:
                dp[i][j] = min(dp[i-1][j],dp[i][j-1],dp[i-1][j-1])+1
    
    return dp[m][n]
```

# 题目**221**✨✨

```python
from typing import List
```

```python
def maximalSquare(self, matrix: List[List[str]]) -> int:
    m = len(matrix)
    n = len(matrix[0])
    dp = [[0] *n for _ in range(m)]
    for i in range(m):
        if matrix[i][0] =="1":
            dp[i][0] = 1
    for j in range(n):
        if matrix[0][j] =="1":
            dp[0][j] = 1
    maxLen = 0
    for i in range(1,m):
        for j in range(1,n):
            if matrix[i][j] =="1":
                if i!=0 and j!=0:
                    dp[i][j] = min(dp[i-1][j],dp[i-1][j-1],dp[i][j-1])+1
                maxLen = max(maxLen,dp[i][j])
    return maxLen*maxLen
    
            
```
