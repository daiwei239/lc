# 题目**1680**✨✨

```python
def concatenatedBinary(self, n: int) -> int:
    res = ""
    for i in range(1,n+1):
        res+=bin(i)
    return int(res,2)
```

# 题目**136**✨

```python
def singleNumber(self, nums: List[int]) -> int:
    res = 0
    for i in nums:
        res^=i
    return res
```
