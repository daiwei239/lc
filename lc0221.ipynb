# 题目**762**✨

```python
def countPrimeSetBits(self, left: int, right: int) -> int:
    res = 0
    for i in range(left,right+1):
        temp = bin(i)
        if temp.count("1") in {2,3,5,7,11,13,17,19}:
            res+=1
    return res
```

# 题目**399**✨✨

```python
from typing import List
from collections import defaultdict
```

```python
def calcEquation(self, equations: List[List[str]], values: List[float], queries: List[List[str]]) -> List[float]:
    graph = defaultdict(list)
    cc = dict()
    # 构建邻接表类型的图
    for (x,y),c in zip(equations,values):
        graph[x].append(y)
        graph[y].append(x)
        cc[(x,y)] = c
        cc[(y,x)] = 1/c 
            
    def get_one_res(pair:List[str],curr,visited)->float:
        start,end  = pair
        # 这个是最重要的一个条件
        if start == end:
            return curr
        
        visited.add(start)
        
        if start not in graph:
            return -1.0
        # 这个只是加快剪枝的一个条件
        if end in graph[start]:
            return curr * cc[(start,end)]
        
        for y_temp in graph[start]:

            if y_temp in visited:
                continue

            calculate = curr * cc[(start,y_temp)]
            rr = get_one_res([y_temp,end],calculate,visited)
            if rr != -1.0:
                return rr
            
        return -1.0
    result = []
    for one in queries:
        if one[0] not in graph or one[1] not in graph:
            result.append(-1.0)
        else:
            result.append(get_one_res(one,1.0,set()))
    return result
```

# 题目**207**✨✨

```python
from collections import defaultdict
```

```python
def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
    pre = defaultdict(list)
    
    for x,y in prerequisites:
        pre[x].append(y)
    
    visited = [0]*numCourses

    def orPass(courseNum:int)->bool:
        if visited[courseNum] == 1:
            return False
        if visited[courseNum] == 2:
            return True
        count = 0 
        visited[courseNum]=1
        for i in pre[courseNum]:
            if orPass(i):
                count+=1
            else:
                return False
                
        if count == len(pre[courseNum]):
            visited[courseNum] = 2
            return True
    
        return False
    
    for i in range(numCourses):
        if not orPass(i):
            return False
    
    return True
```

## 方法二
>拓扑排序的方式进行求解

```python
from collections import defaultdict,deque
```

```python
def canFinish(self, numCourses: int, prerequisites: List[List[int]]) -> bool:
    indegree = [0]*numCourses
    ha = defaultdict(list)
    counter = 0
    for x,y in prerequisites:
        ha[y].append(x)
        indegree[x]+=1
    stack = deque()
    for i in range(len(indegree)):
        if indegree[i] == 0:
            stack.append(i)
            counter+=1
    while stack:
        q = stack.popleft()
        for i in ha[q]:
            indegree[i]-=1
            if indegree[i]==0:
                stack.append(i)
                counter+=1
    return counter == numCourses
    

```

```python
def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
    indegree = [0]*numCourses
    ha = defaultdict(list)
    counter = 0
    res = []
    for x,y in prerequisites:
        ha[y].append(x)
        indegree[x]+=1
    stack = deque()
    for i in range(len(indegree)):
        if indegree[i] == 0:
            stack.append(i)
            counter+=1
    while stack:
        q = stack.popleft()
        res.append(q)
        for i in ha[q]:
            indegree[i]-=1
            if indegree[i]==0:
                stack.append(i)
                counter+=1
    return res if  counter == numCourses else []
    
    
```
