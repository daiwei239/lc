# 题目**1886**✨

```python
from typing import List
```

```python
def findRotation(self, mat: List[List[int]], target: List[List[int]]) -> bool:
    n = len(mat)
    for _ in range(4):
        for i in range(n//2):
            for j in range((n+1)//2):
                mat[i][j],mat[j][n-1-i],mat[n-1-i][n-1-j],mat[n-1-j][i]=\
                mat[n-1-j][i],mat[i][j],mat[j][n-1-i],mat[n-1-i][n-1-j]
        if mat == target:
            return True
    return False 
```

# 题目**105**✨✨

```python
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def buildTree(self, preorder: List[int], inorder: List[int]) -> Optional[TreeNode]:
    pre_idx = 0
    keyvalue = {key:val for val,key in enumerate(inorder)}
    def build(left:int,right:int)->Optional[TreeNode]:
        nonlocal pre_idx
        if left>right:
            return None
        root_val = preorder[pre_idx]
        pre_idx+=1

        mid = keyvalue[root_val]
        root_node = TreeNode(val=root_val)
        root_node.left = build(left,mid-1)
        root_node.right = build(mid+1,right)
        return root_node
    return build(0,len(inorder)-1)
```

# 题目**437**✨✨

```python
def pathSum(self, root: Optional[TreeNode], targetSum: int) -> int:
    res = 0
    def start(node:Optional[TreeNode],target:int):
        if not node:
            return
        nonlocal res
        if node.val == target:
            res+=1
        start(node.left,target-node.val)
        start(node.right,target-node.val)
    def dfs(node:Optional[TreeNode]):
        if node is None:
            return
        start(node,targetSum)
        dfs(node.left)
        dfs(node.right)
    dfs(root)

    return res
        
```

## 方法二，更高效的方法

```python
from collections import defaultdict
```

```python

def pathSum(self, root: TreeNode, targetSum: int) -> int:
    prefix = defaultdict()
    prefix[0] = 1

    def dfs(root:TreeNode,curr:int)->int:
        nonlocal prefix
        if not root:
            return 0
        res = 0
        curr +=root.val
        res +=prefix[curr-targetSum]

        prefix[curr]+=1
        res+=dfs(root.left,curr)
        res+=dfs(root.right,curr)
        prefix[curr]-=1
        return res
    return dfs(root,0)
        
```

# 题目**236**✨✨

```python
def lowestCommonAncestor(self, root: 'TreeNode', p: 'TreeNode', q: 'TreeNode') -> 'TreeNode':
    def dfs(node:'TreeNode')->'TreeNode':
        if not node:
            return node
        if node==p or node == q:
            return node
        left = dfs(node.left)
        right = dfs(node.right)
        if left and right:
            return node
        if left:
            return left
        if right:
            return right
        return None
    return dfs(root)
```

# 题目**200**✨✨

```python
def numIslands(self, grid: List[List[str]]) -> int:
    visited = [[False]*n for _ in range(m)]
    m = len(grid)
    n = len(grid[0])
    def dfs(i:int,j:int)->None:
        nonlocal visited
        if i<0 or i>=m or j<0 or j>=n:
            return 
        if not visited[i][j] and grid[i][j] ==1:
            visited[i][j] = True
            for ii,jj in [(-1,0),(1,0),(0,1),(0,-1)]:
                x,y = i+ii,j+jj
                if 0<=x<m and 0<=y<n:
                    dfs(x,y)
        return
    
    counter = 0
    for i in range(m):
        for j in range(n):
            if grid[i][j] ==1 and not visited[i][j]:
                dfs(i,j)
                counter+=1
    return counter
```

# 题目**994**✨✨

```python
from collections import deque
```

```python
def orangesRotting(self, grid: List[List[int]]) -> int:
    m = len(grid)
    n = len(grid[0])
    times = deque()
    for i in range(m):
        for j in range(n):
            if grid[i][j] ==2:
                times.append((i,j,0))
    t=0
    while times:
        i,j,t = times.popleft()
        for ii,jj in [(-1,0),(1,0),(0,-1),(0,1)]:
            x,y = i+ii,j+jj
            if 0<=x<m and 0<=y<n and grid[x][y]==1:
                grid[x][y]=2
                times.append((x,y,t+1))
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 1:
                return -1
            
    return t
    
                
```
