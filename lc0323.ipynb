# 题目**1594**✨✨

```python
def maxProductPath(self, grid: List[List[int]]) -> int:
        m = len(grid)
        n = len(grid[0])
        curr = [[0]*n for _ in range(m)]
        minarr = [[0]*n for _ in range(m)]
        mod = 10**9 + 7
        minarr[0][0]=curr[0][0] = grid[0][0]
        for i in range(1,m):
            minarr[i][0] =  curr[i][0]=curr[i-1][0]*grid[i][0]
        for j in range(1,n):
            minarr[0][j] = curr[0][j] = curr[0][j-1]*grid[0][j]
        for i in range(1,m):
            for j in range(1,n):
                if grid[i][j]>=0:
                    curr[i][j] = max(curr[i-1][j]*grid[i][j],curr[i][j-1]*grid[i][j])
                    minarr[i][j] = min(minarr[i-1][j]*grid[i][j],minarr[i][j-1]*grid[i][j])
                else:
                    curr[i][j] = min(minarr[i-1][j],minarr[i][j-1])*grid[i][j]
                    minarr[i][j] = max(curr[i-1][j],curr[i][j-1])*grid[i][j]
        if curr[m-1][n-1]<0:
            return -1
        return curr[m-1][n-1]%mod
```

# 题目**35**✨

```python
from typing import List
```

```python
def searchInsert(self, nums: List[int], target: int) -> int:
    left,right = 0,len(nums)-1
    if nums[-1]<target:
        return len(nums)
    while left<right:
        mid =left+(right-left)//2
        if nums[mid]<target:
            left = mid+1
        else:
            right = mid
    return left
```

# 题目**74**✨✨

```python
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
    m,n = len(matrix),len(matrix[0])
    left,right = 0,m*n-1
    while left<=right:
        mid = left+(right-left)//2
        x,y = mid//n ,mid%n
        if matrix[x][y] == target:
            return True
        if matrix[x][y]>target:
            right = mid-1
        if matrix[x][y]<target:
            left = mid+1
    return False
```

# 题目**46**✨✨

```python
def permute(self, nums: List[int]) -> List[List[int]]:
    n = len(nums)
    res = []
    curr = []
    visited = [False]*n
    def dfs():
        nonlocal res
        nonlocal curr
        if len(curr) == n:
            res.append(curr[:])
            return 
        for i,v in enumerate(nums):
            if not visited[i]:
                visited[i] =True
                curr.append(v)
                dfs()
                curr.remove(v)
                visited[i]= False


    dfs()
    return res
    
            
```
