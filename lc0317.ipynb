# 题目**1727**✨✨

```python
from typing import List
```

```python
def largestSubmatrix(self, matrix: List[List[int]]) -> int:
    m = len(matrix)
    n = len(matrix[0])
    res = 0
    for i in range(1,m):
        for j in range(n):
            if matrix[i][j] ==1:
                matrix[i][j] = matrix[i-1][j]+1
    
    for i in range(m):
        matrix[i].sort(reverse=True)
        for j in range(n):
            res = max(res,(j+1)*matrix[i][j])
    return res
```

# 题目**1**✨

```python
def twoSum(self, nums: List[int], target: int) -> List[int]:
    hash = dict()
    for i,val in enumerate(nums):
        if target-val not in hash:
            hash[val] = i
        else:
            return [hash[target-val],i]
```

```python
def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
    hash = dict(list)
    for i in strs:
        temp = i.sort()
        if temp not in hash:
            hash[temp] = i
        else:
            hash[temp].append(i)
    # res = []
    # for v in hash.values():
    #     res.append(v)
    # return res
    return list(hash.values())
```

# 题目**128**✨✨

```python
def longestConsecutive(self, nums: List[int]) -> int:
    nums = set(nums)
    res = 0
    for num in nums:
        if num-1 not in nums:
            curr = 1
            while num+1 in nums:
                num+=1
                curr+=1
            res = max(res,curr)
    return res
```

# 题目**11**✨✨

```python
def maxArea(self, height: List[int]) -> int:
    leftpoint,rightpoint = 0,len(height)-1
    res = 0
    while leftpoint<rightpoint:
        if height[leftpoint]<height[rightpoint]:
            res = max(res,height[leftpoint]*(rightpoint-leftpoint))
            leftpoint+=1
        else:
            res = max(res,height[rightpoint]*(rightpoint-leftpoint))
            rightpoint-=1
    return res
```

# 题目**15**✨✨

```python
def threeSum(self, nums: list[int]) -> list[list[int]]:
    res = []
    nums.sort()
    for i,target in enumerate(nums):
        if i > 0 and nums[i] == nums[i - 1]:   # 修改2：外层去重
            continue
        hash = set()
        for j,v in enumerate(nums[i+1:]):
            
            if -target-v in hash:
                tri = tuple(sorted((target,-target-v,v)))
                res.add(tri)
            hash.add(v)

    return res
```

# 题目**42**✨✨✨

```python
def trap(self, height: List[int]) -> int:
    leftMax =[-inf]
    rightMax = [-inf]
    temp = 0
    n = len(height)
    for i in range(n):
        leftMax.append(max(leftMax[-1],height[i]))
    leftMax = leftMax[1:]
    for j in range(n-1,-1,-1):
        rightMax.append(max(rightMax[-1],height[j]))
    
    rightMax.reverse()

    res = 0

    for i in range(1,n-1):
        t = min(leftMax[i],rightMax[i])
        if t>height[i]:
            res+=t-height[i]
    return res


```
