# 题目**3714**✨✨

```python
def longestBalanced(self, s: str) -> int:
    n = len(s)
    res = 0
    # 第一种情况，只需要一个字符
    last = 1
    for i in range(0,n-1):
        if s[i] ==s[i+1]:
            last+=1
        else:
            res = max(res,last)
            last = 1
    res = max(res,last)
    # 第二种情况，需要两个字符
    def twoEle(x:str,y:str)->int:
        ans = 0
        start = 0
        while start<n:
            if s[start]!=x and s[start]!=y:
                start+=1
                continue
            hash = { 0: start-1}
            diff = 0
            while start<n and (s[start]==x or s[start]==y):
                if s[start]==x:
                    diff+=1
                else:
                    diff-=1
                if diff in hash:
                    ans = max(ans,start-hash[diff])
                else:
                    hash[diff]= start
                start+=1
        return ans
        
    res = max(res,twoEle('a','b'),twoEle('b','c'),twoEle('c','a'))

    # 三个元素字符

    curr =[0,0,0]
    hash = {(0,0):-1}
    for i,ss in enumerate(s):
        curr[ord(ss)-97]+=1
        key = (curr[0]-curr[1],curr[1]-curr[2])
        if key in hash:
            res = max(res,i-hash[key])
        else:
            hash[key]=i
    return res
```

# 题目**222**✨

```python
from collections import Optional
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def countNodes(self, root: Optional[TreeNode]) -> int:
    cnt = 0
    def count(root: Optional[TreeNode]):
        nonlocal cnt
        if root is None:
            return 
        cnt+=1
        count(root.left)
        count(root.right)
    count(root)
    return cnt
        
```

# 题目**199**✨✨

```python
from typing import List
from collections import deque
```

```python
def rightSideView(self, root: Optional[TreeNode]) -> List[int]:
    if root is None:
        return []
    te = deque()
    cnt = 0
    res = list()
    
    te.append(root)
    while te:
        size = len(te)
        cnt = 0
        while cnt <size:
            temp = te.popleft()
            cnt+=1 #弹出去一个就+1
            if size==cnt :
                res.append(temp.val)
            if temp.left:
                te.append(temp.left)
            if temp.right:
                te.append(temp.right)
    return res
        


```

# 题目**637**✨

```python
def averageOfLevels(self, root: Optional[TreeNode]) -> List[float]:
    te = deque()
    res = list()
    te.append(root)
    while te:
        size = len(te)
        cnt = 0
        layerTotal = 0
        while cnt<size:
            temp = te.popleft()
            cnt += 1
            layerTotal += temp.val
            if temp.left:
                te.append(temp.left)
            if temp.right:
                te.append(temp.right)
        res.append(layerTotal/size)
    return res
```

# 题目**102**✨✨

```python
def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
    if root is None:
        return []
    de = deque()
    res = list()
    de.append(root)
    
    while de:
        size = len(de)
        cnt = 0
        currArr=[]
        while cnt < size:
            temp = de.popleft()
            cnt+=1
            currArr.append(temp.val)
            if temp.left:
                de.append(temp.left)
            if temp.right:
                de.append(temp.right)
        res.append(currArr)
    return res

    
```

# 题目**103**✨✨

```python
def zigzagLevelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        if root is None:
            return []
        de = deque()
        res = list()
        de.append(root)
        orReverse = 1
        while de:
            size = len(de)
            cnt = 0
            currArr=deque()
            while cnt < size:
                temp = de.popleft()
                cnt+=1
                if orReverse==1:
                    currArr.append(temp.val)
                else:
                    currArr.appendleft(temp.val)
                if temp.left:
                    de.append(temp.left)
                if temp.right:
                    de.append(temp.right)
            
            res.append(list(currArr))
            orReverse = -orReverse
        return res
```

# 题目**530**✨

```python
def getMinimumDifference(self, root: Optional[TreeNode]) -> int:
    last = float("-inf")
    res = float("inf")
    def inorder(root:Optional[TreeNode]):
        nonlocal res
        nonlocal last
        if root is None:
            return 
        inorder(root.left)
        res = min(res,root.val-last)
        last = root.val
        inorder(root.right)
    inorder(root)
    return res
```
