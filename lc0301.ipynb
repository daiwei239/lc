# 题目**1689**✨✨

```python
def minPartitions(self, n: str) -> int:
    res = 0
    for i in n:
        res = max(res,ord(i)-crd('0'))
    return res
```

# 题目**137**✨✨

```python
from typing import List
```

```python
def singleNumber(self, nums: List[int]) -> int:
    res = 0
    for i in range(32):
        curr = sum((num>>i) &1 for num in nums)
        if curr %3:
            if i == 31:
                res -= 1<<i
            else:
                res |= 1<<i
    return res
```

# 题目**9**✨

```python
def isPalindrome(self, x: int) -> bool:
    if x<0:
        return False
    temp = str(x)
    left,right = 0,len(temp)-1
    while left<right:
        if temp[left] != temp[right]:
            return False
        left+=1
        right-=1
    return True
```
