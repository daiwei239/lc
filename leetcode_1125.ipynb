```python
# 62
def uniquePaths(self, m: int, n: int) -> int:
    res=[[1 for _ in range(n)] for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if i==0 or j ==0:
                continue
            res[i][j]=res[i-1][j]+res[i][j-1]
    return res[m-1][n-1]
```

```python
#34
# 这个题目转变为了，找比当前值小一个的数字和找比当前值大一个的数字
def searchRange(self, nums: List[int], target: int) -> List[int]:
    def binSearch(nums:List[int],target:int):
            l= 0
            r = len(nums)-1
            while(l<=r):
                mid = (l+r)//2
                if nums[mid]<target:
                    l = mid+1
                if nums[mid]>=target:
                    r = mid-1
            return l
    res = [-1,-1]
    le = binSearch(nums,target)
    if le == len(nums) or nums[le]!= target:
        return res
    r = binSearch(nums,target+1)-1
    res[0]=le
    res[1]=r
    return res

```
