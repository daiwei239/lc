# 题目**3567**✨✨

```python
from typing import List
from math import inf
```

```python
def minAbsDiff(self, grid: List[List[int]], k: int) -> List[List[int]]:
    m = len(grid)
    n = len(grid[0])
    res = [[0]*(n-k+1) for _ in range(m-k+1)]
    for i in range(m-k+1):
        for j in range(n-k+1):
            tempList = []
            for ii in range(i,i+k):
                for jj in range(j,j+k):
                    tempList.append(grid[ii][jj])
            tempList.sort()
            minRes = inf
            for i in range(1,k*k):
                if tempList[i]!= tempList[i-1]:
                    minRes = min(minRes,tempList[i]-tempList[i-1])
            if minRes == inf:
                minRes = 0
            res[i][j] = minRes

    return res
```

# 题目**240**✨✨

```python
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
    m = len(matrix)
    n = len(matrix[0])
    
    for j in range(n):
        up,down =0,m-1
        while up<=down:
            mid = up+(down-up)//2
            if matrix[mid][j]<target:
                up = mid+1
            elif matrix[mid][j]>target:
                down = mid-1
            else:
                return True
    return False

        
```

## 方法二，时间复杂度更低的方式

```python
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        m, n = len(matrix), len(matrix[0])
        x, y = 0, n - 1
        while x < m and y >= 0:
            if matrix[x][y] == target:
                return True
            if matrix[x][y] > target:
                y -= 1
            else:
                x += 1
        return False
```

# 题目**142**✨✨

```python
from typing import Optional
```

```python
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None
```

```python
def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
    hash = set()
    while head:
        if head not in hash:
            hash.add(head)
            head = head.next
        else:
            return head
    return None
```

# 题目**21**✨

```python
def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
    dummy = ListNode(-1)
    temp = dummy
    while list1 and list2:
        if list1.val<list2.val:
            temp.next = list1
            list1 = list1.next
        else:
            temp.next = list2
            list2 = list2.next
        temp = temp.next
    temp.next = list1 if list1 else list2
    return dummy.next
```

```python
def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
    count = 0
    dummy = ListNode(-1,head)
    curr = dummy.next
    while curr:
        count+=1
        curr = curr.next
    count=count-n
    curr = dummy
    for _ in range(count):
        curr = curr.next
    if curr.next:
        curr.next = curr.next.next
    return head
```

# 题目**24**✨✨

```python
def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
    if not head or not head.next:
        return head
    dummy = ListNode(-1,head)
    pre = dummy
    curr = pre.next
    nxt = curr.next
    while nxt:
        pre.next = nxt
        curr.next = nxt.next
        nxt.next = curr
        pre = curr
        curr = curr.next 
        nxt = None
        if curr:
            nxt = curr.next
    return dummy.next
```

```python
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
```

```python
def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
    dummy = Node(-1,next=head)
    temp = head
    while temp:
        temp_nxt = Node(temp.val)
        temp_nxt.next = temp.next
        temp.next = temp_nxt
        temp = temp_nxt.next

    temp = head
    while temp:
        if temp.random:
            temp.next.random = temp.random.next
        if temp.next:
            temp = temp.next.next
    
    temp = dummy
    while temp:
        if temp.next:
            temp.next = temp.next.next
        temp = temp.next
    return dummy.next
```

# 题目**146**✨✨

```python
from collections import OrderedDict
```

```python
class DLinkedNode:
    def __init__(self,key = 0,value = 0):
        self.key = key 
        self.val = value
        self.prev = None
        self.next = None

class LRUCache:

    def __init__(self, capacity: int):
        self.cache = dict()
        self.head = DLinkedNode()
        self.tail = DLinkedNode()
        self.head.next = self.tail
        self.tail.prev = self.head
        self.capacity = capacity
        self.size = 0
        

    def get(self, key: int) -> int:
        if key in self.cache:
            node = self.cache[key]
            self.moveToHead(node)
            return node.val
        return -1
        
        
    
    def put(self, key: int, value: int) -> None:
        if key not in self.cache:
            node = DLinkedNode(key,value)
            self.cache[key] = node
            self.addToHead(node)
            self.size+=1
            if self.size>self.capacity:
                removed = self.removeTail()
                self.cache.pop(removed.key)
                self.size -=1
        else:
            node = self.cache[key]
            node.val = value
            self.moveToHead(node)

    def addToHead(self,node:DLinkedNode):
        node.prev = self.head
        node.next = self.head.next
        self.head.next.prev = node
        self.head.next = node

    def removeNode(self,node:DLinkedNode):
        node.prev.next = node.next
        node.next.prev = node.prev

    def moveToHead(self,node:DLinkedNode):
        self.removeNode(node)
        self.addToHead(node)

    def removeTail(self):
        node = self.tail.prev
        self.removeNode(node)
        return node

        
```

# 题目**94**✨

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def inorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = []
        def inorder(node:Optional[TreeNode]):
            nonlocal res
            if node is None:
                return
            inorder(node.left)
            res.append(node.val)
            inorder(node.right)
        inorder(root)
        return res
```

# 题目**104**✨

```python
def maxDepth(self, root: Optional[TreeNode]) -> int:
    def dfs(node:Optional[TreeNode])->int:
        if node is None:
            return 0
        left = dfs(node.left)
        right = dfs(node.right)
        return max(left,right)+1
    return dfs(root)
```
