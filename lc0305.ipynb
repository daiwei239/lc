# 题目**1758**✨✍️

>直接遍历即可  
>交替字符串只有两种，一种是0101 一种是1010  
>这两者各自遍历需要多少次，然后选小的即可

```python
def minOperations(self, s: str) -> int:
    n = len(s)
    if n<2:
        return 0
    
    res1 = 0#从0开头
    res2 = 0#从1开头
    for i in range(n):
        if i % 2 ==0:
            if s[i]=="0":
                res2+=1
            else:
                res1+=1
        
        if i %2 == 1:
            if s[i] =="0":
                res1+=1
            else:
                res2+=1
    return min(res1,res2)
```

# 题目**64**✨✨

```python
from typing import List
from math import inf
```

```python
def minPathSum(self, grid: List[List[int]]) -> int:
    m = len(grid)
    n = len(grid[0])
    dp = [[inf]*n  for _ in range(m)]
    for i in range(m):
        for j in range(n):
            if i ==0 and j ==0:
                continue
            if i ==0:
                dp[i][j] = dp[i][j-1]+grid[i][j]
            elif j == 0:
                dp[i][j] = dp[i-1][j]+grid[i][j]
            else:
                dp[i][j] = min(dp[i-1][j],dp[i][j-1])+grid[i][j]
    return dp[m-1][n-1]
```

# 题目**63**✨✨

```python
def uniquePathsWithObstacles(self, obstacleGrid: List[List[int]]) -> int:
    m = len(obstacleGrid)
    n = len(obstacleGrid[0])
    dp = [[0]*n for _ in range(m)]
    dp[0][0] = 1 if obstacleGrid[0][0]!= 1 else 0

    for i in range(m):
        for j in range(n):
            if i ==0 and j ==0:
                continue
            if obstacleGrid[i][j] ==1:
                dp[i][j] = 0
                continue
            if obstacleGrid[i-1][j]!=1:
                dp[i][j] +=dp[i-1][j]
            if obstacleGrid[i][j-1]!=1:
                dp[i][j] +=dp[i][j-1]
    return dp[m-1][n-1]
```

# 题目**5**✨✨

```python
def longestPalindrome(self, s: str) -> str:
    n = len(str)
    if n<2:
        return n
    dp = [[False]* n for _ in range(n)]
    for i in range(n):
        dp[i][i] = True

    for i in range(0,n):
        for j in range(i+1,n):
            if i+1<=j-1:
                dp[i][j] = dp[i+1][j-1] and s[i] == s[j]
            else:
                dp[i][j] = s[i] == s[j]
    maxlen = 0
    st = 0
    en = 0
    for i in range(n):
        for j in range(n):
            if dp[i][j] == True and j-i+1>maxlen:
                maxlen = j-i+1
                st = i
                en = j
    return s[st:en+1]
    
```
