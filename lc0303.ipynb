# 题目**1545**✨✨✍️

```python
def findKthBit(self, n: int, k: int) -> str:
    if k == 0:
        return "0"
    half = (1<<(n-1))
    if k == half:
        return "1"
    if k<half:
        return findKthBit(n-1,k)
    return "0" if findKthBit(n-1,half*2-k)=="1" else "1"
```

# 题目**779**✨✨

```python
def kthGrammar(self, n: int, k: int) -> int:
    if k == 1 or n==1:
        return 0
    pivot = 1<<(n-2)
    if k<= pivot:
        return self.kthGrammar(n-1,k)
    return 1 if self.kthGrammar(n-1,k-pivot)==0 else 0
```

# 题目**70**✨

```python
def climbStairs(self, n: int) -> int:
    if n ==1:
        return 1
    if n == 2:
        return 2
    minosOne = 2
    minosTwo = 1

    for i in range(3,n+1):
        temp = minosOne
        minosOne +=minosTwo
        minosTwo = temp
    return minosOne
```

# 题目**198**✨✨

```python
from typing import List
```

```python
def rob(self, nums: List[int]) -> int:
    if len(nums)==1:
        return nums[0]
    dp = [0]*len(nums)
    dp[0] = nums[0]
    dp[1] = max(nums[0],nums[1])
    for i in range(2,len(nums)):
        dp[i] = max(dp[i-1],dp[i-2]+nums[i])
    return dp[len(nums)-1]
```

## 方法二、用常数的内存空间做

```python
def rob(self, nums: List[int]) -> int:
    n = len(nums)
    if n ==1:
        return nums[0]
    llast = nums[0]
    last = max(nums[0],nums[1])
    for i in range(2,n):
        temp = last
        last = max(llast+nums[i],last)
        llast = temp
    return last
```
