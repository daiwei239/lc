# 题目**1888**✨✨✍️

```python
def minFlips(self, s: str) -> int:
    n = len(s)
    s= s+s
    count = 0
    ans = n
    for j in range(len(s)):
        if int(s[j])%2==j%2:
            count+=1
        i = j-n+1
        if i<0:
            continue
        ans = min(ans,count,n-count)
        if  int(s[i])%2 == i%2:
            count-=1
    return ans
```

# 题目**53**✨✨

```python
from typing import List
from typing import inf
```

```python
def maxSubArray(self, nums: List[int]) -> int:
    minSum = 0
    res = -inf
    curr = 0
    for i in range(len(nums)):
        curr = curr+nums[i]
        res = max(res,curr-minSum)
        minSum = min(minSum,curr)
    return res
```

# 题目**73**✨✨

```python
def setZeroes(self, matrix: List[List[int]]) -> None:
    rows = set()
    cols = set()
    m = len(matrix)
    n = len(matrix[0])
    for i in range(m):
        for j in range(n):
            if matrix[i][j]==0:
                rows.add(i)
                cols.add(j)
    
    for i in range(m):
        for j in range(n):
            if i in rows or j in cols:
                matrix[i][j]= 0
```

# 题目**173**✨✨

```python
from typing import Optional
```

```python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class BSTIterator:

    def __init__(self, root: Optional[TreeNode]):
        self.queue = []
        self.index = -1
        self.inorder(root)
    def inorder(root:Optional[TreeNode]):
        if root is None:
            return
        if root.left:
            inorder(root.left)
        self.queue.append(root.val)
        if root.right:
            inorder(root.right)

    def next(self) -> int:
        self.index+=1
        return self.queue[self.index]
        

    def hasNext(self) -> bool:
        if self.index <len(self.queue):
            return True
        return False
        
```

# 题目**123**✨✨✨

```python
def maxProfit(self, prices: List[int]) -> int:
    k = 2
    dp = [[0,-inf] for _ in range(k+1)]
    for price in prices:
        for j in range(k,0,-1):
            dp[j][0] = max(dp[j][0],dp[j][1]+price)
            dp[j][1] = max(dp[j][1],dp[j-1][0]-price)
    return dp[k][0]
```

# 题目**918**✨✨

```python
from collections import deque
```

```python
def maxSubarraySumCircular(self, nums: List[int]) -> int:
    n = len(nums)
    q = deque()
    pre,res = nums[0],nums[0]
    q.append((0,pre))
    for i in range(1,2*n):
        while q and q[0][0]<i-n:
            q.popleft()
        pre+=nums[i%n]
        res = max(res,pre-q[0][1])
        while q and q[-1][1]>=pre:
            q.pop()
        q.append((i,pre))
    return res
```

```python
def maxSubarraySumCircular(self, nums: List[int]) -> int:
    n = len(nums)
    nums = nums+nums
    maxVal = 0
    for i in range(n):
        curr = 0
        minVal = 0
        for j in range(i,i+n):
            curr +=nums[j]
            maxVal = max(maxVal,curr-minVal)
            minVal = min(minVal,curr)
    return maxVal



```
