题目**2976**

```python
from typing import List
```

>本题指的是多个节点到多个节点的最短路径是多少，  
>可以直接采用floyd算法求解
>
>

```python
def minimumCost(self, source: str, target: str, original: List[str], changed: List[str], cost: List[int]) -> int:
    G = [[inf] *26 for _ in range(26)]
    for i in range(26):
        G[i][i] = 0
    
    # 初始化
    for x,y,z in zip(original,changed,cost):
        idx = ord(x)-ord("a")
        idy = ord(y)-ord("a")
        G[idx][idy]= min(G[idx][idy],z)
    
    # 进行floyd算法进行求解不同原点到不同点的距离
    for k in range(26):
        for i in range(26):
            for j in range(26):
                G[i][j] = min(G[i][j],G[i][k]+G[k][j])
    # 正式求解
    ans = 0

    for x,y in zip(source,target):
        if x==y:
            continue
        idx = ord(x)-ord("a")
        idy = ord(y)-ord("a")
        if G[idx][idy]==inf:
            return -1
        ans +=G[idx][idy]

    return ans

```

题目**3112**

```python
from heapq import *
```

```python
def minimumTime(self, n: int, edges: List[List[int]], disappear: List[int]) -> List[int]:
    g = [[] for _ in range(n)]
    inf = float("inf")
    dist = [inf]*n
    h = [(0,0)]
    for x,y,wt in edges:
        g[x].append((y,wt))#这个地方不需要进行判断，
        # 这个判断只能获得一个残缺的范围，
        # 既不是充分条件，也不是必要条件，没有判断必要
    
    while h:
        curr_dis,x=heappop(h)
        if curr_dis>dist[x]:
            continue
        #*
        if curr_dis>disappear[x]:
            continue
        #*
        for old_dis,y in g[x]:

            # if curr_dis>dist[y]:
            #     continue
            new_dis = curr_dis+old_dis
            if new_dis<dist[y]:
                if new_dis<disappear[y]:
                    dist[y] = new_dis
                    heappush(h,(new_dis,y))
                # else:
                #     dist[y]=inf
                #     new_dis = inf
                # heappush(h,(inf,y))
    for i in range(len(dist)):
        if dist[i]==inf:
            dist[i] =-1
    return dist
```

题目**3604**

```python
def minTime(self, n: int, edges: List[List[int]]) -> int:
    g = [[] for _ in range(n)]
    for x,y,start,end in edges:
        g[x].append((y,start,end))
    dist=[inf]*n
    dist[0] = 0
    h=[(0,0)]

    while h:
        dx,x = heappop(h)
        if dx>dist[x]:
            continue
        if x ==n-1:
            return dx
        for y,start,end in g[x]:
            endtime = max(dx,start)+1
            # 在[start,end]区间内走就可以了
            if endtime-1<=end and endtime<dist[y]:
                dist[y]=endtime
                heappush(h,(dist[y],y))
    return -1
```

题目**169**
>直接用hash表去做即可
>
>
>

```python
def majorityElement(self, nums: List[int]) -> int:
    hashtable = dict()
    for i in nums:
        hashtable[i]=hashtable.get(i,0)+1
    return max(hashtable.items(),key= lambda x:x[1])[0]
```
