# 题目**761** ✨✨✨

>将子串进行分割处理  
>将其按照长度进行逆序排序  
>返回整体的串

```python
def makeLargestSpecial(self, s: str) -> str:
    if len(s)<=2:
        return s
    diff = 0
    sub = []
    start = 0
    for i,ch in enumerate(s):
        if ch =="1":
            diff+=1
        else:
            diff-=1
            if diff == 0:
                sub.append("1"+ self.makeLargestSpecial(s[start+1:i])+"0")
                start=i+1

    sub.sort(reverse=True)
    return "".join(sub)
```

# 题目**35**✨

```python
from typing import List
```

```python
def searchInsert(self, nums: List[int], target: int) -> int:
    left,right = 0,len(nums)-1
    while left<=right:
        mid = (left+right)//2
        if nums[mid]<target:
            left = mid+1
        elif nums[mid]>target:
            right = mid - 1
        else:
            return mid
    return left
```
