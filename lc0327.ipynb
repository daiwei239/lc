```python
from typing import List
```

```python
def areSimilar(self, mat: List[List[int]], k: int) -> bool:
    m = len(mat)
    n = len(mat[0])
    for i in range(m):
        for j in range(n):
            if i %2 ==0:
                if mat[i][j]!=mat[i][(j+k)%n]:
                    return False
            else:
                if mat[i][j]!= mat[i][(j-k+n)%n]:
                    return False
    
    return True
```

# 题目**739**✨✨

```python
def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
    n = len(temperatures)
    answer = [0]*n
    stack = []
    stack.append(0)
    for i in range(1,n):
        while stack and temperatures[i]>temperatures[stack[-1]]:
            day = stack.pop()
            answer[day] = i-day
        stack.append(i)
    return answer            
```

# 题目**215**✨✨

```python
from random import randint


def findKthLargest(self, nums: List[int], k: int) -> int:
    n = len(nums)
    target = n-k
    left ,right = 0,n-1
    while left<right:
        l = left
        r = right
        pivot = randint(left, right)
        pivot_val = nums[pivot]
        while l<=r:
            while l<= r and nums[l]<=pivot_val:
                l+=1
            while l<= r and nums[r]>pivot_val:
                r-=1
            if l<=r:
                nums[l],nums[r] = nums[r],nums[l]
                l+=1
                r-=1
        if target<=r:
            right = r
        elif target>=l:
            left = l
        else:
            return nums[target]
    return nums[left]
```

# 题目**55**✨✨

```python
def canJump(self, nums: List[int]) -> bool:
    maxlen = nums[0]
    i = 0
    while i<maxlen:
        maxlen = max(maxlen,nums[i]+i)
        if maxlen>=len(nums)-1:
            return True
        i+=1
    return False
    
```

# 题目**45**✨✨

```python
def jump(self, nums: List[int]) -> int:
    steps = 0
    maxlen = 0
    i = 0
    while maxlen<len(nums)-1:
        temp = maxlen
        
        while i<=temp:
            maxlen = max(maxlen,i+nums[i])
            i+=1
        steps+=1
    return steps
```

# 题目**118**✨

```python
def generate(self, numRows: int) -> List[List[int]]:
    res = []
    
    if numRows==1:
        return [[1]]
    for row in range(numRows):
        curr = []
        for i in range(row+1):
            if i==0 or i== row:
                curr.append(1)
            else:
                curr.append(res[row-1][i-1]+res[row-1][i])
        res.append(curr)
    return res
```

# 题目**198**✨✨

```python
def rob(self, nums: List[int]) -> int:
    n = len(nums)
    dp = [0]*(n+1)
    dp[0] = 0
    dp[1] = nums[0]
    for i in range(2,n+1):
        dp[i] = max(dp[i-1],dp[i-2]+nums[i-1])
    return dp[n]
    
        
```

# 题目**18**✨

```python
def isPalindrome(self, s: str) -> bool:
        if not s:
            return True
        s = s.strip()
        s = s.lower()
        left,right = 0,len(s)-1
        while left<=right:
            if not s[left].isdigit() and not s[left].isalpha():
                left+=1
                continue
            if not s[right].isdigit() and not s[right].isalpha():
                right-=1
                continue
            if s[left]!=s[right]:
                return False
            left+=1
            right-=1
        return True
```

```python
def minimumTotal(self, triangle: List[List[int]]) -> int:
    m = len(triangle)
    dp = []
    for i in range(1,m+1):
        dp.append([0]*i)
    dp[0][0] = triangle[0][0]
    for i in range(2,m+1):
        for j in range(i):
            if j ==0:
                dp[i-1][j] = triangle[i-1][j]+dp[i-2][0]
                continue
            if  j ==i-1:
                dp[i-1][j] = triangle[i-1][j]+dp[i-2][j-1]
                continue
            dp[i-1][j] = min(dp[i-2][j-1],dp[i-2][j])+triangle[i-1][j]
    return min(dp[m-1])
```
