# 题目**1009**✨✍️

```python
def bitwiseComplement(self, n: int) -> int:
    if n == 0:
        return 1
    mask = (1<<n.bit_length())-1
    return n^mask
```

```python
def findComplement(self, num: int) -> int:
        if num == 0:
            return 1
        mask = (1<<num.bit_length())-1
        return num^mask
```

# 题目**427**✨✨

```python
from typing import List
class Node:
    def __init__(self, val, isLeaf, topLeft, topRight, bottomLeft, bottomRight):
        self.val = val
        self.isLeaf = isLeaf
        self.topLeft = topLeft
        self.topRight = topRight
        self.bottomLeft = bottomLeft
        self.bottomRight = bottomRight
```

```python
def construct(self, grid: List[List[int]]) -> 'Node':
    def dfs(left:int,right:int,top:int,down:int)->'Node':
        if left == right and top == down:
            return Node(
                val=grid[left][top],
                isLeaf=True,
                topLeft=None,
                topRight=None,
                bottomLeft=None,
                bottomRight=None
            )
        midlr = left+(right-left)//2
        midtd = top+(down-top)//2
        tl = dfs(left,midlr,top,midtd)
        tr = dfs(left, midlr, midtd+1, down)  
        bl = dfs(midlr+1,right,top,midtd)
        br = dfs(midlr+1,right,midtd+1,down)
        
        if tl.isLeaf and tr.isLeaf and bl.isLeaf and br.isLeaf and tl.val==tr.val==bl.val==br.val:
            return Node(
                val = tl.val,
                isLeaf=1,
                topLeft=None,
                topRight=None,
                bottomLeft=None,
                bottomRight=None
            )
        return Node(
            val= 1,
            isLeaf=False,
            topLeft=tl,
            topRight=tr,
            bottomLeft=bl,
            bottomRight=br
        )
    m = len(grid)-1
    n = len(grid[0])-1
    return dfs(0,m,0,n)
```

# 题目**295**✨✨✨✅

```python
import heapq
```

```python
class MedianFinder:

    def __init__(self):
        # 使用left作为最大堆，每次输出最大的出去
        self.left = []
        # 使用right作为最小堆，每次输出最小的出去
        self.right = []
        

        

    def addNum(self, num: int) -> None:
        # 如果当前left和right数量相同，就添加到left
        # 如果left数量多，且当前值小于left的最大值，也添加到left，
        # # 然后将其最大值弹出传递给right
        if len(self.left)==len(self.right):
            if self.right and num>self.right[0]:
                temp = heapq.heappushpop(self.right,num)
                heapq.heappush(self.left,-temp)
            else:
                heapq.heappush(self.left,-num)
        else:
            if num<-self.left[0]:
                temp= -heapq.heappushpop(self.left,-num)
                heapq.heappush(self.right,temp)
            else:
                heapq.heappush(self.right,num)
        
        
        

    def findMedian(self) -> float:
        if len(self.left)==len(self.right):
            return (-self.left[0]+self.right[0])/2
        return -self.left[0]
    
```

# 题目**224**✨✨✨

>两者的加减是在遇到下一个符号的时候才进行的，当遇到下一个运算符的时候，  
>会将上一个运算符给

```python
def calculate(self, s: str) -> int:
    res,num,sign = 0,0,1
    stack = []
    for c in s:
        if c.isdigit():
            num = 10 * num+int(c)
        elif c == "+" or c =="-":
            res += sign *num
            num = 0
            sign = 1 if c == "+" else -1
        elif c =="(":
            stack.append(res)
            stack.append(sign)
            res = 0
            sign = 1
        elif c ==")":
            res +=sign*num
            num = 0
            res *=stack.pop()
            res += stack.pop()

    res += sign*num
    return res
```
