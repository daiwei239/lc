```python
from typing import List
```

```python
def sortByBits(self, arr: List[int]) -> List[int]:
    def get_count(x:int):
        res = 0
        while x:
            res +=x&1
            x = x>>1
        return res
    bit_counts = [0] * 10001
    for x in arr:
        bit_counts[x] = get_count(x)

    arr.sort(key=lambda x:(bit_counts[x],x))
    return arr
```

# 题目**74**✨✨

```python
def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
    m,n = len(matrix),len(matrix[0])
    left,right = 0,m*n-1
    while left<=right:
        mid = (left+right)//2
        x = matrix[mid // n][mid % n]
        if x == target:
            return True
        if x < target:
            left = mid+1
        else:
            right = mid-1
    return False

```
