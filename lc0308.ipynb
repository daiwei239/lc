# 题目**1980**✨✨

```python
from typing import List
```

```python
def findDifferentBinaryString(self, nums: List[str]) -> str:
    n = len(nums[0])
    nums_set = set(nums)
    for i in range(2**n):
        if bin(i)[2:].zfill(n) not in nums_set:
            return bin(i)[2:].zfill(n)
```

# 题目**Q2**

```python
def smallestBalancedIndex(self, nums: list[int]) -> int:
    n = len(nums)
    if n ==1:
        return -1
    leftRes = [0]*n
    rightRes = [1]*n
    

    for i in range(1,n):
        
        leftRes[i] = leftRes[i-1]+nums[i-1]
    for j in range(n-2,-1,-1):
        
        rightRes[j] = rightRes[j+1]*nums[j+1]
    
    for i in range(n):
        if leftRes[i] == rightRes[i]:
            return i
    return -1

```

## 方法二，内存不爆炸算法

```python
def smallestBalancedIndex(self, nums: list[int]) -> int:
    total_sum = sum(nums)
    right_sum = 0
    right_prod = 1
    res = -1
    
    LIMIT = 10**16 # 设定一个超过所有可能累加和的上限
    
    for i in range(len(nums) - 1, -1, -1):
        val = nums[i]
        left_sum = total_sum - right_sum - val
        
        if left_sum == right_prod:
            res = i
            
        right_sum += val
        right_prod *= val
        
        if right_prod > LIMIT:
            right_prod = LIMIT + 7
            
    return res
```

# 题目Q3

```python
def minOperations(self, s: str) -> int:
    s_sort = "".join(sorted(s))
    if len(s)==2 and s[0]!= s_sort[0]:
        return -1
    if s == s_sort:
        return 0
    if s[0]==s_sort[0]:
        return 1
    if s[0] == s_sort[-1] and s[-1]==s_sort[0]:
        mid = s[1:-1]
        if s[0] not in mid and s[-1] not in mid:
            return 3
    return 2
    
```

# 题目Q4

```python
def minCost(self, s: str, encCost: int, flatCost: int) -> int:
    def divide(s:str):
        X = s.count("1")
        if len(s) ==1:
            if X == 0:
                return flatCost
            return encCost
        L = len(s)


        if X == 0:
             return flatCost
        if L%2!= 0: 
            return L * X * encCost
        
        s1= s[0:L//2]
        s2 = s[L//2:]
        cost1 = L * X * encCost
        cost2 = divide(s1)+divide(s2)
        return min(cost1,cost2)
        
    if s.count("1") == 0:
           return flatCost
    return divide(s)
        
```

# 题目**201**✨✨

```python
def rangeBitwiseAnd(self, left: int, right: int) -> int:
    shift = 0
    while left<right:
        left>>=1
        right>>=1
        shift+=1
    return left<<shift
```

# 题目**373**✨✨

```python
from collections import heappop,heappush
```

```python
def kSmallestPairs(self, nums1: List[int], nums2: List[int], k: int) -> List[List[int]]:
    m,n = len(nums1),len(nums2)
    ans = []
    pq = [(nums1[i]+nums2[0],i,0) for i in range(min(k,m))]
    while pq and len(ans)<k:
        _,i,j = heappop(pq)
        ans.append((i,j))
        if j+1<n:
            heappush((nums1[i]+nums2[j+1],i,j+1))
    
    return ans
```

# 题目**42**✨✨✨——接雨水

```python
def trap(self, height: List[int]) -> int:
    n = len(height)
    left = [height[0]]+[0]*(n-1)
    leftMax = height[0]

    for i in range(1,n):
        leftMax = max(leftMax,height[i])
        left[i]= leftMax
    
    right = [0]*(n-1)+[height[-1]]
    rightMax = height[-1]

    for j in range(n-2,-1,-1):
        rightMax = max(rightMax,height[j])
        right[j] = rightMax
    
    res = sum(min(left[i],right[i])-height[i] for i in range(1,n-1))
    return res        
```

## 方法二——用双指针的方式去求解，空间复杂度为常数

```python
def trap(self, height: List[int]) -> int:
    left,right = 0,len(height)-1
    leftMax,rightMax = 0,0
    ans = 0
    while left<right:
        leftMax = max(leftMax,height[left])
        rightMax = max(rightMax,height[right])
        if rightMax>leftMax:
            ans += leftMax-height[left]
            left+=1
        else:
            ans += rightMax-height[right]
            right-=1
    return ans
```

# 题目**135**✨✨✨——分糖果

```python
def candy(self, ratings: List[int]) -> int:
    
    n = len(ratings)
    candies = [1]*n
    for i in range(1,n):
        if ratings[i]>ratings[i-1]:
            candies[i]= candies[i-1]+1
    for j in range(n-2,-1,-1):
        if ratings[j]>ratings[j+1]:
            candies[j] = max(candies[j],candies[j+1]+1)
    
    return sum(candies)

```

# 题目**124**✨✨✨

```python
from typing import Optional
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def maxPathSum(self, root: Optional[TreeNode]) -> int:
    ans = -inf
    def dfs(root:Optional[TreeNode])->int:
        nonlocal ans
        if root is None:
            return 0
        
        leftVal = max(dfs(root.left),0)
        rightVal = max(dfs(root.right),0)
        ans = max(ans,leftVal+rightVal+root.val)
        return root.val+max(leftVal,rightVal)
        
    dfs(root)
    return ans
```

```python
def maxPathSum(self, root: Optional[TreeNode]) -> int:
    ans = -inf
    def dfs(root:Optional[TreeNode])->int:
        nonlocal ans
        if root is None:
            return 0
        
        leftVal = max(dfs(root.left),0)
        rightVal = max(dfs(root.right),0)
        if root.val>0:
            ans = max(ans,leftVal+rightVal+root.val)
        else:
            ans = max(ans,leftVal,rightVal)
        return root.val+max(leftVal,rightVal)
        
    dfs(root)
    return ans
```
